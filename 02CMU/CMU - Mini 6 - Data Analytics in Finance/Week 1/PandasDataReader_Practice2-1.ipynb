```python
#ng = DataReader('DHHNGSP', 'fred', start='2004-1-1', end='2019-12-31')
#ng.to_csv('data/ng.csv') 


ng = pd.read_csv('data/ng.csv', parse_dates=['DATE'], index_col='DATE')
ng.columns = ['price']
ng = ng.dropna()
ng.tail()
```

```python
ng.mean()
```

```python
ng_rolling = ng.rolling(window=30).mean()
ng_rolling
```

```python
print(ng.var())
print(ng_rolling.var())
```

```python
rolling_high = ng['price'].rolling(window = 30).max()
ind = ng['price'] <  rolling_high
ind
```

```python
ng_filtered = ng['price'][ind]
ng_filtered.shape
#ng_filtered.plot()
```
