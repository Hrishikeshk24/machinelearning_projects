### **Installing DataReader**
* To install a package, type conda install in the terminal window: 
    * conda install -c anaconda pandas-datareader
* -c denotes the channel to download from 
    * https://repo.anaconda.com/pkgs/main 
    * Another useful channel: conda-forge 
    * conda install -c conda-forge package_name

* The package is installed into the current environment
    * An environment contains a collection of compatible packages. 
    * With environments, we can maintain multiple collections of compatible packages.
