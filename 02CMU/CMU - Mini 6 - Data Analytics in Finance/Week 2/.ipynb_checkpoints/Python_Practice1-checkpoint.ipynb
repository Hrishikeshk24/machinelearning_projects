```python
from pandas_datareader.data import DataReader
from datetime import date, datetime, timedelta
import pandas as pd
import numpy as np
```

```python
ticker = 'BP'
source = 'yahoo'
end_date = date(2019,12,31)
begin_date = date(2004,1,1)
#df_BP = DataReader(ticker, source,  begin_date, end_date)
from pandas_datareader import data as DataReader
import yfinance as yf
yf.pdr_override()
df_BP = DataReader.get_data_yahoo(ticker, start=begin_date, end=end_date)

#df_BP.to_csv('data/df_BP.csv') 

#df_BP = pd.read_csv('data/df_BP.csv', parse_dates=['Date'], index_col='Date')
```

```python
print(type(df_BP))
print(df_BP.shape)
print(df_BP.info())
```

```python
volume = df_BP['Volume']
type(volume)
```

```python
print(df_BP.iloc[0:3,:])
print(df_BP.iloc[-3:,:])
```

```python
df_BP['MKTCap'] = 0 # Broadcasts to entire column
df_BP['MKTCap'][::3] = 1000 # here the step size is 3 
df_BP.head(5)
```

```python
df1 = df_BP['Adj Close'].to_frame()
df2 = df_BP['High'].to_frame()
df4 = df2.join(df1)
df3 = pd.concat([df2, df1], axis='rows')
print(df3)
print(df4)
```

```python
df_BP.loc['2019-02-19']
df_BP.loc['2019-02-19',['High','Low']]
df_BP.loc['2019-02',['High','Low']]
df_BP.loc['2019-02':'2019-03',['High','Low']]
```

```python
df_BP_monthly = df_BP.resample('M').last()
df_BP_monthly.head()
```
