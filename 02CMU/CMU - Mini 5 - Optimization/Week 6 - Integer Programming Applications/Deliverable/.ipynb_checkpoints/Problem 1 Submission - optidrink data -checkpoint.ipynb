```python
# data
Cost = {'c1' : 1.4, 'c2' : 1.7, 'c3' : 0.8, 'c4' : 2, 'c5' : 1.6, 'c6' : 1.5, 'c7' : 1.1, 'c8' : 0.9}

Content = {
    ('c1','Energy') : 300, 	('c1','Flavor1') : 0.4, 	('c1','Flavor2') : 0.6, 	
    ('c2','Energy') : 200, 	('c2','Flavor1') : 0.6, 	('c2','Flavor2') : 0.5, 	
    ('c3','Energy') : 510, 	('c3','Flavor1') : 0.4, 	('c3','Flavor2') : 0.3, 	
    ('c4','Energy') : 400, 	('c4','Flavor1') : 0.7, 	('c4','Flavor2') : 0.4, 	
    ('c5','Energy') : 360, 	('c5','Flavor1') : 0.5, 	('c5','Flavor2') : 0.7, 	
    ('c6','Energy') : 150, 	('c6','Flavor1') : 0.3, 	('c6','Flavor2') : 0.3, 	
    ('c7','Energy') : 240, 	('c7','Flavor1') : 0.3, 	('c7','Flavor2') : 0.2, 	
    ('c8','Energy') : 350, 	('c8','Flavor1') : 0.5, 	('c8','Flavor2') : 0.6
}

MinReq = {'Energy' : 400, 'Flavor1' : 0.5, 'Flavor2' : 0.4}
MaxReq = {'Energy' : 500, 'Flavor1' : 0.55, 'Flavor2' : 0.45}
```

```python
Ingredients = Cost.keys()
```

```python
Ingredients
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
select = mdl.binary_var_dict(Ingredients, name='select')
amount = mdl.continuous_var_dict(Ingredients, name='amount')
```

```python
# objective
mdl.minimize(mdl.sum(Cost[i]*amount[i] for i in Ingredients))
```

```python
# constraint: the total content of c1-c8 should add up to one unit

mdl.add_constraint(mdl.sum(amount[i] for i in Ingredients) == 1 )
```

```python
# constraint: it must contain between 400 and 500 kJ

mdl.add_constraint(mdl.sum((Content[i,j]-MinReq[j])*amount[i] for (i,j) in Content.keys() if j == 'Energy') >= 0)
```

```python
mdl.add_constraint(mdl.sum((Content[i,j]-MaxReq[j])*amount[i] for (i,j) in Content.keys() if j == 'Energy') <= 0)
```

```python
# constraint: it must contain between 0.5 and 0.55 units of Flavor1

mdl.add_constraint(mdl.sum((Content[i,j]-MinReq[j])*amount[i] for (i,j) in Content.keys() if j == 'Flavor1') >= 0)
```

```python
mdl.add_constraint(mdl.sum((Content[i,j]-MaxReq[j])*amount[i] for (i,j) in Content.keys() if j == 'Flavor1') <= 0)
```

```python
# constraint: it must contain between 0.4 and 0.45 units of Flavor2

mdl.add_constraint(mdl.sum((Content[i,j]-MinReq[j])*amount[i] for (i,j) in Content.keys() if j == 'Flavor2') >= 0)
```

```python
mdl.add_constraint(mdl.sum((Content[i,j]-MaxReq[j])*amount[i] for (i,j) in Content.keys() if j == 'Flavor2') <= 0)
```

```python
# constraint: If we choose to select a specific ingredient,the mix must contain at least 0.1 and at most 0.4 units 
# of that ingredient.

for i in Ingredients:
    mdl.add_constraint(amount[i] <= 0.4*select[i])
    mdl.add_constraint(amount[i] >= 0.1*select[i])
```

```python
# new constraint:

Rules = { 'R1' }
Combination = { 
    ('R1','c1') : 0, ('R1','c2') : 0, ('R1','c3') : 1, ('R1','c4') : 1
}
```

```python
RuleConstraints = {}
for r in Rules:
    # sum the 1s
    SumOnes = mdl.sum(select[i] for i in select if (r,i) in Combination if Combination[r,i] == 1)
    # sum the 0s
    SumZeroes = mdl.sum((1-select[i]) for i in select if (r,i) in Combination if Combination[r,i] == 0)
    RuleConstraints[r] = mdl.add_constraint(SumOnes + SumZeroes <= 3)
```

```python
for r in Rules:
    print(RuleConstraints[r])
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```
