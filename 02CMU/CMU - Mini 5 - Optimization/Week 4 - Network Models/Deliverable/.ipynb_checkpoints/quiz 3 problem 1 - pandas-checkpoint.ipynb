```python
# Read in the csv file as a dataframe using pandas
import pandas as pd
df = pd.read_csv("quiz 3 problem 1.csv", index_col='from')

# Initialize the set of nodes, and the dictionaries for cost and demand/supply:
Nodes = set()
Cost = {}
Supply = {}

for i in df.index:  # iterate over the row index
    for j in df:    # iterative over the columns
        if df[j][i] != '-':
            if i == 'Demand':
                Supply[j] = float(df[j][i])
            elif j == 'Capacity':
                Supply[i] = -float(df[j][i])
            else:
                Cost[(i,j)] = float(df[j][i])
                Nodes.add(i)
                Nodes.add(j)

# Complete the "Supply" data for the nodes without a supply (i.e., the warehouses have 0 demand/supply).
for i in Nodes:
    if not(i in Supply.keys()):
        Supply[i] = 0
    
# Define the arc set
Arcs = Cost.keys()
```
