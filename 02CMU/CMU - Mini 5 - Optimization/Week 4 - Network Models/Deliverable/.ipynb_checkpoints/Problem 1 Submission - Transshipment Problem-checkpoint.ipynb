<b>Quiz 3 - Problem 1</b>

```python
# data

import csv
f = open("quiz 3 problem 1.csv")
csvfile = csv.DictReader(f, delimiter=',')
headers = csvfile.fieldnames

table = []
for row in csvfile:
    table.append(row)
    
f.close()

# Create set of nodes
Nodes = set()
# start by adding the elements in the header (except for first and last element)
for i in range(1,len(headers)-1):
    Nodes.add(headers[i])

# Create dictionaries "Cost" and "Supply" (for the retail nodes)
# We can borrow the keys from "Cost" to define the arc set of the network.
Cost = {}
Capacity = {}
Supply = {}
for row in table:
    val = row['from']
    if val != 'Demand':
        for i in Nodes:
            if (row[i] != '-'):
                Cost[val,i] = float(row[i])
    else:
        for i in Nodes:
            if (row[i] != '-'):
                Supply[i] = float(row[i])

# Add the "Supply" data for the factories and complete the node set.
for row in table:
    val = row['from']
    if val != 'Demand':
        if row['Capacity'] != '-':
            Supply[val] = -float(row['Capacity'])
        Nodes.add(val)

# Complete the "Supply" data for the nodes without a supply (i.e., the 
# warehouses have 0 demand/supply).
for i in Nodes:
    if not(i in Supply.keys()):
        Supply[i] = 0
    
# Define the arc set
Arcs = Cost.keys()
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
shipped = mdl.continuous_var_dict(Arcs, lb=0, name='shipped')
```

```python
# objective
mdl.minimize(mdl.sum(Cost[i,j]*shipped[i,j] for (i,j) in Arcs))
```

```python
# constraints
for j in Nodes:
    maxF1R1 = mdl.sum(shipped[i,j] for i in Nodes if (i,j) == ('F1', 'R1'))
    mdl.add_constraint(maxF1R1 <= 400)
```

```python
# constraints
for j in Nodes:
    maxF2R4 = mdl.sum(shipped[i,j] for i in Nodes if (i,j) == ('F2', 'R4'))
    mdl.add_constraint(maxF2R4 <= 350)
```

```python
# constraints
for j in Nodes:
    if j == 'W1':
        maxW1 = mdl.sum(shipped[i,j] for i in Nodes if i == 'F1' or i=='F2')
        mdl.add_constraint(maxW1 <= 2000)
```

```python
# constraints
for j in Nodes:
    if j == 'W2':
        maxW2 = mdl.sum(shipped[i,j] for i in Nodes if i == 'F1' or i=='F2')
        mdl.add_constraint(maxW2 <= 2000)
```

```python
# constraints
for j in Nodes:
    inflow = mdl.sum(shipped[i,j] for i in Nodes if (i,j) in Arcs)
    outflow = mdl.sum(shipped[j,i] for i in Nodes if (j,i) in Arcs)
    mdl.add_constraint(inflow - outflow >= Supply[j])
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```
