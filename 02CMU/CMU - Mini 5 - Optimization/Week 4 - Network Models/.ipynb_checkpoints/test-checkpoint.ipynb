```python
# data 
Nodes = range(1,9)
UnitCost = {(1, 8) : 5, (2, 6) : 5, (2, 8) : 4, (3, 7) : 1, (3, 8) : 4, (4, 6) : 1, (4, 8) : 3, (5, 6) : 4, (5, 7) : 5, (5, 8) : 3}
Supply = { 1 : -6, 2 : -6, 3 : -6, 4 : -6, 5 : -6, 6 : 2, 7 : 1, 8 : 2 }
```

```python
Arcs = { (i,j) for (i,j) in UnitCost }
```

```python
Arcs
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# flow variable for each arc
flow = mdl.continuous_var_dict(Arcs, lb=0, name='flow')
```

```python
# objective
mdl.maximize(mdl.sum(flow[i,j]*UnitCost[i,j] for (i,j) in Arcs))
```

```python
# flow balance constraint for each node
for i in Nodes:
    inFlow = mdl.sum(flow[j,i] for j in Nodes if (j,i) in Arcs)
    outFlow = mdl.sum(flow[i,j] for j in Nodes if (i,j) in Arcs)
    mdl.add_constraint(inFlow - outFlow >= Supply[i])
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```
