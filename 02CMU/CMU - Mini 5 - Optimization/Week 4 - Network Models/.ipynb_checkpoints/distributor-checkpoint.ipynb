<b>Distributor Transshipment Problem</b>

```python
# data
Nodes = range(1,7)
Cost = { (1,3) : 20, (1,4) : 30, (2,4) : 10, (2,5) : 20, (3,4) : 20, (3,6) : 40, (4,6) : 30, (5,4) : 50, (5,6) : 30 }
UpperBound = {(1,3) : 100, (1,4) : 10, (2,4) : 100, (2,5) : 100, (3,4) : 100, (3,6) : 10, (4,6) : 100, (5,4) : 100, (5,6) : 100 }
Arcs = { (i,j) for (i,j) in Cost}
Supply = { 1 : -30, 2 : -40, 3 : -20, 4 : 10, 5 : 30, 6 : 40 }
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
shipped = mdl.continuous_var_dict(Arcs, lb=0, ub=UpperBound, name='shipped')
```

```python
# objective
mdl.minimize(mdl.sum(Cost[i,j]*shipped[i,j] for (i,j) in Arcs))
```

```python
# constraints
for j in Nodes:
    inflow = mdl.sum(shipped[i,j] for i in Nodes if (i,j) in Arcs)
    outflow = mdl.sum(shipped[j,i] for i in Nodes if (j,i) in Arcs)
    mdl.add_constraint(inflow - outflow >= Supply[j])
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```
