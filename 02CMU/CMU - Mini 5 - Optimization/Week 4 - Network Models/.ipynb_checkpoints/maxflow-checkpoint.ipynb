<b>Maximum Flow Problem</b>

```python
# data
Nodes = {'Wash, DC', 'Chicago', 'Dallas', 'Kansas City', 'Denver', 'San Fran.' }
Capacity = { ('Wash, DC','Chicago') : 800, ('Wash, DC','Dallas') : 700, ('Wash, DC','Kansas City') : 650, ('Chicago','Dallas') : 725, ('Chicago','Denver') : 700, ('Kansas City','Dallas') : 625, ('Kansas City','Denver') : 750, ('Kansas City','San Fran.') : 500, ('Denver','San Fran.') : 900, ('Dallas','San Fran.') : 725, ('San Fran.','Wash, DC') : 10000000 }
Arcs = { (i,j) for (i,j) in Capacity}
Supply = {'Wash, DC' : 0, 'Chicago' : 0, 'Dallas' : 0, 'Kansas City' : 0, 'Denver' : 0, 'San Fran.' : 0 }
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables, including the capacity constraints as upper bounds
calls = mdl.continuous_var_dict(Arcs, lb=0, ub=Capacity, name='calls')
```

```python
# objective
mdl.maximize(calls['San Fran.', 'Wash, DC'])
```

```python
# flow balance constraints
for j in Nodes:
    inflow = sum(calls[i,j] for i in Nodes if (i,j) in Arcs)
    outflow = sum(calls[j,i] for i in Nodes if (j,i) in Arcs)
    mdl.add_constraint(inflow - outflow == Supply[j])
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```
