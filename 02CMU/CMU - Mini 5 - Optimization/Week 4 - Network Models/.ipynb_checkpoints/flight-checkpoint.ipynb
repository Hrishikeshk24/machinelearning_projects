<b>Shortest Path: Flight Schedule</b>

```python
# data
Nodes = {'DFW', 'HNL', 'LAX', 'LGA', 'MIA', 'ORD', 'PVD', 'SFO' }
Supply = {'DFW' : 0, 'HNL' : 1, 'LAX' : 0, 'LGA' : 0, 'MIA' : 0, 'ORD' : 0, 'PVD' : -1, 'SFO' : 0 }
Distance = {('DFW','LAX') : 1233, ('DFW','LGA') : 1387, ('DFW','MIA') : 1120, ('DFW','ORD') : 802, 
            ('HNL','LAX') : 2555, ('LAX','DFW') : 1233, ('LAX','HNL') : 2555, ('LAX','ORD') : 1743, 
            ('LAX','SFO') : 337, ('LGA','DFW') : 1387, ('LGA','MIA') : 1099, ('LGA','PVD') : 142, 
            ('MIA','DFW') : 1120, ('MIA','LGA') : 1099, ('MIA','PVD') : 1205, ('ORD','DFW') : 802, 
            ('ORD','LAX') : 1743, ('ORD','PVD') : 849, ('ORD','SFO') : 1843, ('PVD','LGA') : 142, 
            ('PVD','MIA') : 1205, ('PVD','ORD') : 849, ('SFO','LAX') : 337, ('SFO','ORD') : 1843 }
Arcs = {(i,j) for (i,j) in Distance}
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
leg = mdl.continuous_var_dict(Arcs, lb=0, ub=1, name='leg')
```

```python
# objective
mdl.minimize(mdl.sum(Distance[i,j]*leg[i,j] for (i,j) in Arcs))
```

```python
# flow conservation constraint
for j in Nodes:
    inflow = mdl.sum(leg[i,j] for i in Nodes if (i,j) in Arcs)
    outflow = mdl.sum(leg[j,i] for i in Nodes if (j,i) in Arcs)
    mdl.add_constraint(inflow - outflow == Supply[j])
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```
