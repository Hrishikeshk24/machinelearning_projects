```python
# Read in the csv file as a dataframe using pandas
import pandas as pd

df = pd.read_csv("Performance.csv", index_col='Consultant')
Performance = {}
for i in df.index:           # iterate over the row index (consultants)
    for j in df.columns:     # iterate over the columns (projects)
        if df[j][i] != 0:    # entries with performance score = 0 cannnot be allocated
            Performance[i,j] = df[j][i]

df = pd.read_csv("ProjectInfo.csv", index_col='Project')
Required = {}
for j in df.index:           # iterate over the row index (projects)
    Required[j] = df['Required'][j]
            
Projects = Required.keys()
Consultants = { i for (i,j) in Performance.keys() }
```

```python
df = pd.read_csv("SalaryCost.csv", index_col='Consultant')
Cost = {}
for j in df.index:           # iterate over the row index (projects)
    Cost[j] = df['Cost'][j]
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
select = mdl.binary_var_matrix(Consultants, Projects, name='select')
```

```python
# new variables
consultant = mdl.binary_var_dict(Consultants, name='consultant')
```

```python
# new Objective
mdl.minimize(mdl.sum(consultant[i]*Cost[i] for i in Consultants))
```

```python
# Demand Constraints for each Project

for i in Projects:
    mdl.add_constraint(mdl.sum(select[j,i] for j in Consultants if (j,i) in Performance.keys()) == Required[i])
```

```python
# Each consultant can be assigned to at most six projects.

for i in Consultants:
    mdl.add_constraint(mdl.sum(select[i,j] for j in Projects if (i,j) in Performance.keys()) <= 6)
```
