```python
Time1_30 = range(1,31)
```

```python
maintenance_day = [20, 50, 80, 110, 140, 170, 200, 230, 260, 290, 320, 350]
```

```python
len(maintenance_day)
```

```python
from docplex.mp.model import Model
mdl = Model()
mdl.parameters.mip.tolerances.mipgap = 0.000000001
```

```python
# variables
Produced = mdl.continuous_var_dict(TimeHorizon, lb=0, ub=ProductionCapacity, name='Production in a day')
```

```python
# auxiliary variables: inventory at end of month
Inventory = mdl.continuous_var_dict(TimeHorizon, lb=0, ub=StorageCapacity, name='Inventory at end of each day')
```

```python
# New binary variable to represent maintenance day
maintenance = mdl.binary_var_dict(TimeHorizon, name='maintenance')
```

```python
# objective: minimize cost
TotalProdCost = mdl.sum(UnitProdCost[i]*Produced[i] for i in TimeHorizon)
TotalHoldingCost = mdl.sum(UnitStorageCost*Inventory[i] for i in TimeHorizon)
mdl.minimize(TotalProdCost + TotalHoldingCost + 450* len(maintenance_day))

mdl.add_kpi(TotalProdCost, 'Total production cost')
mdl.add_kpi(TotalHoldingCost, 'Total holding cost')
```

```python
# inventory balance constraints
for i in TimeHorizon:
    if i==1:
        mdl.add_constraint(Inventory[1] == Produced[1] - Demand[1])
    else:
        mdl.add_constraint(Inventory[i] == Inventory[i-1] + Produced[i] - Demand[i])
```

```python
#at least one maintenance day is planned for days 1, 2, . . . , 30.

mdl.add_constraint(mdl.sum(maintenance[i] for i in Time1_30) >= 1)
```

```python
#at least one maintenance day is planned for every sequence of 30 consecutive days

mdl.add_constraint(mdl.sum(maintenance[i] for i in TimeHorizon if i in maintenance_day) >= 12)
```

```python
# no production is possible on day of maintenance.
mdl.add_constraint(mdl.sum(Produced[i] for i in TimeHorizon if i in maintenance_day) ==0)
```

```python
# solve 
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```
