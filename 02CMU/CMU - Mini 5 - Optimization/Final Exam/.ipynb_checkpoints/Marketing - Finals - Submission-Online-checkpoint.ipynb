```python
Price = {'P1':35, 'P2':45, 'P3': 28}
Campaigns = range(0,11)
Products = Price.keys()

# import math for e
import math

# Pre-compute demand for each product i and campaign j:
Demand = {}
for j in Campaigns:
        Demand['P1',j] = 500*(1+math.log(1+j,10))
        Demand['P2',j] = 350*(1+math.log(1+2*j,10))
        Demand['P3',j] = 600*(1+math.log(1+1.1*j,10))
```

```python
Cost = {'P1':1000, 'P2':1250, 'P3':850}
```

```python
Products
```

```python
Demand
```

```python
# pre-compute the revenue for each product i and campaign j 
Revenue = {}
for i in Products:
    for j in Campaigns:
        if Demand[i,j] > 2500:
            Revenue[i,j] = Price[i]*2500
        else:
            Revenue[i,j] = Price[i]*Demand[i,j]
```

```python
Revenue
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
assign = mdl.binary_var_matrix(Products, Campaigns, name='assign')
```

```python
# objective
mdl.maximize(mdl.sum(Revenue[i,j]*assign[i,j] for (i,j) in assign.keys()))
```

```python
# constraint: each product has one specific number of campaigns
for i in Products:
    mdl.add_constraint(mdl.sum(assign[i,j] for j in Campaigns) == 1)
```

```python
# New Constraint:total cost of the marketing plan is at most $15,000.

mdl.add_constraint(mdl.sum(j*assign[i,j]*Cost[i] for (i,j) in assign.keys())<=15000)
```

```python
# New Constraint: P1 has strictly more marketing campaigns than each of P2 and P3.
        
campaigns_P2 = mdl.sum(j*assign[i,j] for (i,j) in assign.keys() if i == 'P2')
campaigns_P3 = mdl.sum(j*assign[i,j] for (i,j) in assign.keys() if i == 'P3')
campaigns_P1 = mdl.sum(j*assign[i,j] for (i,j) in assign.keys() if i == 'P1')
mdl.add_constraint(campaigns_P1 >= campaigns_P2 + campaigns_P3)
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```
