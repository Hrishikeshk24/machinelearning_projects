```python
import pandas as pd
df_twitter = pd.read_csv("problem2.csv")

Price = {}
Follows = {}    # Follows[i,j] = 1 if follower j follows influencer i
for i in df_twitter.index:
    Price[df_twitter['Influencer'][i]] = df_twitter['Price'][i]
    for j in range(1,91):
        follower = "F"+str(j)
        Follows[(df_twitter['Influencer'][i],follower)] = df_twitter[follower][i]

Followers = { "F"+str(j) for j in range(1,91) }
```

```python
# import docplex
from docplex.mp.model import Model
mdl = Model()
```

```python
Influencers = Price.keys()

# variable: select influencers
Select = mdl.binary_var_dict(Influencers, name='Select')

# variable: is follower covered?
IsCovered = mdl.binary_var_dict(Followers, name='IsCovered')
```

```python
# objective: maximize the number of followers reached
Reached = mdl.sum(IsCovered[j]  for j in IsCovered)
mdl.maximize(Reached)
mdl.add_kpi(Reached,'Number of  followers reached ')
```

```python
# constraint: budget is at most 30,000
Budget = 30000
mdl.add_constraint(mdl.sum(Select[i]*Price[i] for i in Select) <= Budget, "budget")
```

```python
# constraint: follower can only be reached if it follows a selected influencer
for j in Followers:
    mdl.add_constraint(mdl.sum(Select[i] for i in Influencers if Follows[i,j]==1) >= IsCovered[j])
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```

```python
mdl.report()
```

```python
#online part

# variable representing whether user j follows at least two influencers
FollowsTwo = mdl.binary_var_dict(Followers, name='follows at least two')
```

```python
# updated objective 
TotalReached = mdl.sum(IsCovered[j]+0.4*FollowsTwo[j] for j in IsCovered)
mdl.maximize(TotalReached)
```

```python
# linking constraint: following at least two influencers
for j in Followers:
    mdl.add_constraint(mdl.sum(Select[i] for i in Influencers if Follows[i,j]==1) >= 2*FollowsTwo[j])
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```

```python
# response curve
# To plot the curve I create two lists, for the X (budget) and Y (overall impact) data.
CurveX = []
CurveY = []

for r in range(1,100):
    # set the RHS of the minimum return constraint
    mdl.get_constraint_by_name("budget").rhs = r*1000
    mdl.solve()
    CurveX.append(r*1000)
    CurveY.append(TotalReached.solution_value)
```

```python
import matplotlib.pyplot as plt

plt.plot(CurveX, CurveY)
plt.show()
```
