```python
import pandas as pd
df_problem3a = pd.read_csv("problem3a.csv").set_index('Ad')
df_problem3b = pd.read_csv("problem3b.csv").set_index('Slot')
```

```python
Available = {}
for i in df_problem3b.index:
    Available[i] = df_problem3b['Available'][i]
```

```python
Slots = Available.keys()
Number = {}
ClickThrough = {}
for i in df_problem3a.index:
    Number[i] = df_problem3a['Number'][i]
    for j in Slots:
        ClickThrough[(i,j)] = df_problem3a[j][i]
```

```python
# Optimization model
from docplex.mp.model import Model
mdl = Model()
```

```python
# Variables
Ads = Number.keys()
allocate = mdl.continuous_var_matrix(Ads, Slots, lb=0, name="allocate")
```

```python
# Objective
mdl.maximize(mdl.sum(allocate[i,j]*ClickThrough[i,j] for (i,j) in allocate))
```

```python
# Allocate all orders
for i in Ads:
    mdl.add_constraint(mdl.sum(allocate[i,j] for j in Slots) == Number[i])
```

```python
# Do not exceed available impressions for each slot
for j in Slots:
    mdl.add_constraint(mdl.sum(allocate[i,j] for i in Ads) <= Available[j])
```

```python
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```

```python
# online part:
# additional data
df_problem3c = pd.read_csv("problem3c.csv").set_index('Ad')

MinResponse = {}
Category = {}
for i in df_problem3c.index:
    MinResponse[i] = df_problem3c['MinResponse (x 0.001)'][i]
    Category[i] = df_problem3c['Category'][i]
Categories = set(Category.values())
```

```python
# additional variables and constraints
InSlot = mdl.binary_var_matrix(Ads, Slots, name="InSlot")

# minimum and maximum allocation
for (i,j) in InSlot:
    mdl.add_constraint(allocate[i,j] >= 0.1*Number[i]*InSlot[i,j])
    mdl.add_constraint(allocate[i,j] <= 0.5*Number[i]*InSlot[i,j])
```

```python
# Minimum response rate
for i in Ads:
    mdl.add_constraint(mdl.sum(allocate[i,j]*ClickThrough[i,j] for j in Slots) >= MinResponse[i])
```

```python
# Optimum with default tolerance: 36821118.276 (may be different depending on your machine specifications)
# Optional: A higher precision will result in a better answer: objective 36821576.897
#mdl.parameters.mip.tolerances.mipgap.set(float(0.00000001))
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```

```python
# second set of constraints (online part)
# Every ad with at least 600,000 impressions must be placed in at least three slots.
for i in Ads:
    if Number[i] >= 600000:
        mdl.add(mdl.sum(InSlot[i,j] for j in Slots) >= 3)
```

```python
# No two ads from the same category can be placed in the same slot.
for j in Slots:
    for c in Categories:
        if c != 'Other':
            mdl.add_constraint(mdl.sum(InSlot[i,j] for i in Ads if Category[i]==c) <= 1)
```

```python
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```
