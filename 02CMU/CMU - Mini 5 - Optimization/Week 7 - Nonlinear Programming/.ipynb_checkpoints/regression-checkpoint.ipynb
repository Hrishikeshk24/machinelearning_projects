```python
import csv
f = open("regression.csv")
csvfile = csv.DictReader(f, delimiter='\t')

headers = csvfile.fieldnames

table = []
for row in csvfile:
    table.append(row)
    
f.close()
```

```python
print(headers)
```

```python
for row in table:
    print(row)
```

```python
Observations = {}
for row in table:
    Observations[float(row['x'])] = float(row['y'])
```

```python
print(Observations)
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
# fit quadratic function y = ax^2 + bx + c
#
# CPLEX has lower bound of 0 by default
# -infinity: -mdl.infinity
a = mdl.continuous_var(lb=-mdl.infinity, name='a')
b = mdl.continuous_var(lb=-mdl.infinity, name='b')
c = mdl.continuous_var(lb=-mdl.infinity, name='c')
```

```python
# objective
SumSquaredError = mdl.sum((a*x**2 + b*x + c - Observations[x])**2 for x in Observations.keys())
mdl.minimize(SumSquaredError)
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```
