```python
# import math for constant e
import math

# pre-compute the revenue for each product i and price value v
Revenue = {}
for i in Products:
    for v in Values:
        Revenue[i,v] = v*a[i]*math.e**(b[i]*v)
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
assign = mdl.binary_var_matrix(Products, Values, name='assign')
```

```python
# objective
mdl.minimize(mdl.sum(v*assign[i,v] for (i,v) in assign.keys()))
```

```python
# constraint: each product gets 1 price value
for i in Products:
    mdl.add_constraint(mdl.sum(assign[i,v] for v in Values) == 1)
```

```python
# constraint: minimum revenue
mdl.add_constraint(mdl.sum(Revenue[i,v]*assign[i,v] for (i,v) in assign.keys()) >= MinRevenue)
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```
