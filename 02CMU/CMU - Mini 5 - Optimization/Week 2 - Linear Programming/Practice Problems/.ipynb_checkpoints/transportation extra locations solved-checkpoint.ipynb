```python
# Recall the transportation problem and linear programming solution below
# We add an extra plant and customer, and we add one of the new constraints from the problem set
```

```python
# Transportation Problem: data
Plants = ['Plant 1', 'Plant 2', 'Plant 3']
Customers = ['Cust A', 'Cust B', 'Cust C', 'Cust D']

Cost = {
    ('Plant 1', 'Cust A') : 15,
    ('Plant 1', 'Cust B') : 35,
    ('Plant 1', 'Cust C') : 25,
    ('Plant 1', 'Cust D') : 25,
    ('Plant 2', 'Cust A') : 10,
    ('Plant 2', 'Cust B') : 30,
    ('Plant 2', 'Cust C') : 40,
    ('Plant 2', 'Cust D') : 40,
    ('Plant 3', 'Cust A') : 20,
    ('Plant 3', 'Cust B') : 25,
    ('Plant 3', 'Cust C') : 30,
    ('Plant 3', 'Cust D') : 30
}

Supply = {
    'Plant 1' : 70,
    'Plant 2' : 80,
    'Plant 3' : 50
}

Demand = {
    'Cust A' : 30,
    'Cust B' : 30,
    'Cust C' : 30,
    'Cust D' : 30
}
```

```python
from docplex.mp.model import Model
from docplex.util.environment import get_environment
mdl = Model()
```

```python
# for this example we use a 2-dimensional matrix of variables: 
shipped = mdl.continuous_var_matrix(Plants, Customers, lb=0, name='shipped')        
```

```python
# the objective is to minimize the total weighted shipping costs:
objective = mdl.sum(shipped[(p,c)]*Cost[(p,c)] for p in Plants for c in Customers)
mdl.minimize(objective)
```

```python
# define the demand constraints:
for c in Customers:
    mdl.add_constraint(mdl.sum(shipped[(p,c)] for p in Plants) == Demand[c])
```

```python
# define the supply constraints:
for p in Plants:
    mdl.add_constraint(mdl.sum(shipped[(p,c)] for c in Customers) <= Supply[p])
```

```python
# new constraint that the amount shipped from Plant 2 must be at least 50% of the amount shipped from Plant 1
mdl.add_constraint(mdl.sum(shipped[('Plant 2',c)] for c in Customers) >= 0.5 * mdl.sum(shipped[('Plant 1',c)] for c in Customers))


```

```python
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```
