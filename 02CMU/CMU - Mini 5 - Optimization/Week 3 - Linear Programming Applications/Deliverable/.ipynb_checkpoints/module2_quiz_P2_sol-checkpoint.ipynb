```python
import csv

# read demand data
f = open("module2_quiz_P2a.csv")
csvfile = csv.DictReader(f)
headers = csvfile.fieldnames

table = []
for row in csvfile:
    table.append(row)
    
f.close()

Groups = []
Demand = {}
Months = range(1,6)
for row in table:
    Groups.append(row['Demand (tonne)'])
    for m in Months:
        Demand[(row['Demand (tonne)'], m)] = float(row[str("Month "+str(m))])

# read cost data
f = open("module2_quiz_P2b.csv")
csvfile = csv.DictReader(f)
headers = csvfile.fieldnames

table = []
for row in csvfile:
    table.append(row)
    
f.close()
                
Cost = {}
for row in table:
    for m in Months:
        Cost[(row['Production cost ($/tonne)'], m)] = float(row[str("Month "+str(m))])
        
# other input data
ProductionCapacity = 175
StorageCapacity = 50
UnitStorageCost = 50

# Original 
#InitialInventory = {'Large' : 0, 'Medium' : 5.3, 'Small' : 0}

# Online quiz
InitialInventory = {'Large' : 2.3, 'Medium' : 5.3, 'Small' : 3.7}
```

```python
print(Groups)
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
GroupMonth = Cost.keys()
produced = mdl.continuous_var_dict(GroupMonth, lb=0, name='produced')
storage = mdl.continuous_var_dict(GroupMonth, lb=0, name='storage')
```

```python
# objective
productionCost = mdl.sum(Cost[i,t]*produced[i,t] for (i,t) in GroupMonth)
storageCost = mdl.sum(UnitStorageCost*storage[i,t] for (i,t) in GroupMonth)
totalCost = productionCost + storageCost
mdl.minimize(totalCost)
mdl.add_kpi(productionCost, "Production cost")
mdl.add_kpi(storageCost, "Storage cost")
mdl.add_kpi(totalCost, "Total cost")
```

```python
# Inventory balance constraints

# Month 1:
for i in Groups:
    mdl.add_constraint(InitialInventory[i] + produced[i,1] == Demand[i,1] + storage[i,1])
# Months 2..5: 
for t in range(2,6):
    for i in Groups:
        mdl.add_constraint(0.96*storage[i,t-1] + produced[i,t] == Demand[i,t] + storage[i,t])
```

```python
# Production capacity constraints
for t in Months:
    mdl.add_constraint(mdl.sum(produced[i,t] for i in Groups) <= ProductionCapacity)
```

```python
# Storage capacity constraints
for t in Months:
    mdl.add_constraint(mdl.sum(storage[i,t] for i in Groups) <= StorageCapacity)
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```

```python
mdl.report()
```

```python
for g in Groups:
    for m in Months:
        print(str(g) + " month " + str(m) + ": " + str(produced[g,m].solution_value))
```
