```python
# read csv data file
import csv
f = open("module2_quiz_P1.csv")
csvfile = csv.DictReader(f)

headers = csvfile.fieldnames

table = []
for row in csvfile:
    table.append(row)
    
f.close()
```

```python
# define set of Segments
Segments = []
for row in table:
    if row['Segment'] != 'Cost per ad':
        Segments.append(row['Segment'])
        
# define set of Outlets
Outlets = headers[:]
Outlets.remove('Segment')
Outlets.remove('Minimum Impressions')
```

```python
# define parameters
Cost = {}
for o in Outlets:
    Cost[o] = float(table[-1][o])

Demand = {}
for row in table:
    if row['Segment'] != 'Cost per ad':
        Demand[row['Segment']] = float(row['Minimum Impressions'])
        
Impressions = {}
for row in table:
    if row['Segment'] != 'Cost per ad':
        for o in Outlets:
            Impressions[(row['Segment'],o)] = float(row[o])
```

```python
# build optimization model
from docplex.mp.model import Model
mdl = Model()
```

```python
# variables
amount = mdl.continuous_var_dict(Outlets, lb=0, name= 'amount')
```

```python
# objective
mdl.minimize(mdl.sum(Cost[o]*amount[o] for o in Outlets))
```

```python
# constraints
for s in Segments:
    mdl.add_constraint(mdl.sum(Impressions[s,o]*amount[o] for o in Outlets) >= Demand[s])
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```

```python
# additional constraint: no more than 10% of total spending allocated to web and mobile
#
# This version reflects the 15% requirement of the online quiz.
mdl.add_constraint(Cost['W']*amount['W'] + Cost['M']*amount['M'] <= 
                   0.15*(mdl.sum(Cost[o]*amount[o] for o in Outlets)))
```

```python
# solve
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```

```python
mdl.add_kpi(Cost['M']*amount['M'], "Mobile costs")
```

```python
mdl.report()
```
