```python
import csv

# read demand data
f = open("module2_quiz_P2a.csv")
csvfile = csv.DictReader(f)
headers = csvfile.fieldnames

table = []
for row in csvfile:
    table.append(row)
    
f.close()

Groups = []
Demand = {}
Months = range(1,6)
for row in table:
    Groups.append(row['Demand (tonne)'])
    for m in Months:
        Demand[(row['Demand (tonne)'], m)] = float(row[str("Month "+str(m))])

# read cost data
f = open("module2_quiz_P2b.csv")
csvfile = csv.DictReader(f)
headers = csvfile.fieldnames

table = []
for row in csvfile:
    table.append(row)
    
f.close()
                
Cost = {}
for row in table:
    for m in Months:
        Cost[(row['Production cost ($/tonne)'], m)] = float(row[str("Month "+str(m))])
```

```python
# Maximum Production
ProductionCapacity = 175

# Maximum Inventory
HoldingCost = 50
StorageCapacity = 50
initial_storage = {"Large": 0, "Medium": 5.3, "Small": 0}
storage_rot = 0.04
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# decision variables
Produced = mdl.continuous_var_matrix(Groups, Months, lb = 0, name='Production in month')
```

```python
# auxiliary variables
Inventory = mdl.continuous_var_matrix(Groups, Months, lb = 0, name='Inventory at end of month')
```

```python
# objective function 
TotalProdCost = mdl.sum(Cost[g,m]*Produced[g,m] for g in Groups for m in Months)
TotalHoldingCost = mdl.sum(HoldingCost*Inventory[g,m] for g in Groups for m in Months)
mdl.minimize(TotalProdCost + TotalHoldingCost)
mdl.add_kpi(TotalProdCost, 'Total production cost')
mdl.add_kpi(TotalHoldingCost, 'Total holding cost')
```

```python
for m in Months:
    mdl.add_constraint(mdl.sum(Produced[g, m] for g in Groups) <= ProductionCapacity)
    mdl.add_constraint(mdl.sum(Inventory[g, m] for g in Groups) <= StorageCapacity)


for g in Groups:
    for m in Months:
        if m == 1:
            mdl.add_constraint(Inventory[g, m] == initial_storage [g] + Produced[g, m] - Demand[g, m])
        else:
            mdl.add_constraint(Inventory[g, m] == Inventory[g, m-1]* (1-storage_rot) + Produced[g, m] - Demand[g, m])
```

```python
# solve 
mdl.solve()
mdl.get_solve_details()
```

```python
mdl.print_solution()
```

```python
mdl.report()
```
