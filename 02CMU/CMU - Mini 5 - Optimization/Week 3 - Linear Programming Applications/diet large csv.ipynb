```python
import csv
f = open("diet_large.csv")
csvfile = csv.DictReader(f)

headers = csvfile.fieldnames
```

```python
print(headers)
```

```python
table = []
for row in csvfile:
    table.append(row)
    
f.close()
```

```python
Foods = []
for row in table:
    if row['Foods'] != 'n/a':
        Foods.append(row['Foods'])
```

```python
Nutrients = headers[:]
Nutrients.remove('Foods')
Nutrients.remove('Price/Serving')
Nutrients.remove('Serving Size')
Nutrients.remove('n/a')
```

```python
Cost = {}
for row in table:
    if row['Foods'] != 'n/a':
        Cost[row['Foods']] = float(row['Price/Serving'])
```

```python
Nutrition = {}
for row in table:
    if row['Foods'] != 'n/a':
        for n in Nutrients:
            Nutrition[(row['Foods'], n)] = float(row[n])
```

```python
Types = {}
Requirements = {}
for n in Nutrients:
    Types[n] = table[-1][n]
    Requirements[n] = table[-2][n]
```

```python
print(Types)
```

```python
print(Requirements)
```

```python
# Select fields for output
SelectedFields = ['Foods', 'Serving Size', 'Calories']
for r in table:
    rowout = {}
    for s in SelectedFields:
        rowout[s] = r[s]
    print(rowout)
```

```python
# Write selected fields to output file
SelectedFields = ['Foods', 'Serving Size', 'Calories']     # select the fields to write
f = open("SelectedFields.csv", 'w', newline='')            # open the file for output
csvfileout = csv.DictWriter(f, fieldnames=SelectedFields)  # csvfileout is a reference to the DictWriter
csvfileout.writeheader()                                   # write the header
for r in table:                                            # for each row in the table
    rowout = {}                                            # create a dictionary
    for s in SelectedFields:                               # for each of the selected fields
        rowout[s] = r[s]                                   # add "s : r[s]" to rowout
    csvfileout.writerow(rowout)                            # write the row to the file
f.close()
```

```python
# alternative method that closes the file automatically
with open("SelectedFields2.csv", 'w', newline='') as f:
    SelectedFields = ['Foods', 'Serving Size', 'Calories']
    csvfileout = csv.DictWriter(f, fieldnames=SelectedFields)
    csvfileout.writeheader()
    for r in table:
        rowout = {}
        for s in SelectedFields:
            rowout[s] = r[s]
        csvfileout.writerow(rowout)
```
