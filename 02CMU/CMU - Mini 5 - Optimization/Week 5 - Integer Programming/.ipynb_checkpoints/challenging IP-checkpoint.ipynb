```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# define variables
AllVariables = {}
for i in range(1,13):
    AllVariables[i] = mdl.continuous_var()
for i in range(13,63):
    AllVariables[i] = mdl.binary_var()
# fix even-indexed continuous variables to zero
for i in range(2,13,2):
    mdl.add_constraint(AllVariables[i] == 0)
```

```python
# objective
mdl.minimize(mdl.sum(AllVariables[i]*Profit[i] for i in Profit))
```

```python
# knapsack constraints
for j in Knapsacks:
    mdl.add_constraint(mdl.sum(AllVariables[i]*Weight[i,j] for i in AllVariables if (i,j) in Weight) == Budget[j])
```

```python
# model details
mdl.print_information()
```

```python
# solve
mdl.parameters.timelimit = 10
mdl.solve()
mdl.get_solve_details()
```
