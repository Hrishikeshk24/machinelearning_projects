<b>Example Branch-and-Bound</b>

```python
# Project selection problem
Projects = range(1,7)
Cost = { 1 : 28, 2 : 17, 3 : 26, 4 : 32, 5 : 18, 6 : 24}
Return = { 1 : 58, 2 : 39, 3 : 56, 4 : 90, 5 : 45, 6 : 67 }
Budget = 100
```

```python
from docplex.mp.model import Model
mdl = Model()
```

```python
# for LP relaxation: continuous variables
select = mdl.continuous_var_dict(Projects, lb=0, ub=1, name='select')
```

```python
# objective
mdl.maximize(mdl.sum(select[i]*Return[i] for i in Projects))
```

```python
# knapsack constraint
mdl.add_constraint(mdl.sum(select[i]*Cost[i] for i in Projects) <= Budget)
```

```python
# solve
mdl.solve()
```
