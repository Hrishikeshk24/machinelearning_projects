```python
from docplex.mp.model import Model
```

```python
mdl = Model()
```

```python
x = mdl.continuous_var(name="wheat")
y = mdl.continuous_var(name="corn")
z = mdl.continuous_var(name="potatoes")
```

```python
mdl.add_constraint(3*x + 2*y + 3*z <= 150) # Labor constraint
```

```python
mdl.add_constraint(2*x + 4*y +3*z <= 120) # fertilizer constraint
```

```python
mdl.add_constraint(x + y + z <= 60) #acre constraint
```

```python
mdl.maximize(200*x + 300*y + 275*z)
```

```python
mdl.add_constraint(x >= 0)
```

```python
mdl.add_constraint(y >= 0)
```

```python
mdl.add_constraint(z >= 0)
```

```python
mdl.solve()
```

```python
mdl.print_solution()
```

```python
mdl.print_information()
```

```python
mdl.report()
```
