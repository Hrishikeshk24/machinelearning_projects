```python
from docplex.mp.model import Model
```

```python
mdl = Model()
```

```python
wheat = mdl.continuous_var()
corn = mdl.continuous_var()
```

```python
mdl.add_constraint(wheat + corn <= 60)
```

```python
mdl.add_constraint(3*wheat + 2*corn <= 100)
```

```python
mdl.add_constraint(2*wheat + 4*corn <= 120)
```

```python
mdl.maximize(200*wheat + 300*corn)
```

```python
mdl.print_information()
```

```python
mdl.add_constraint(wheat >= 0)
mdl.add_constraint(corn >= 0)
```

```python
mdl.solve()
```

```python
mdl.report()
```
