```python
# Importing required libraries/packages

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining required variables

aircraft_capacity = 189
standard_ticket_price = 249
premium_ticket_price = 349
std_ticket_mean = 150
std_ticket_stdev = 30
premium_ticket_mean = 100
premium_ticket_stdev = 40
trials = 2
```

```python
tot_revenue =[]
rng = np.random.default_rng(seed=123)
for i in range(trials):
    
    standard_ticket_demand = rng.normal(loc= std_ticket_mean, scale = std_ticket_stdev)
    revenue_standard = int(standard_ticket_demand) * standard_ticket_price
    print("Number of Std tickets:", int(standard_ticket_demand))
    print("revenue from std tickets :", revenue_standard)
    
    premium_ticket_demand = rng.normal(loc = premium_ticket_mean, scale = premium_ticket_stdev)
    print("Premium ticket demand from rng:", int(premium_ticket_demand))
    count = int(standard_ticket_demand)
    print("count:", count)   
    revenue_premium = 0
    for demand in range(1,int(premium_ticket_demand) + 1):
        revenue_premium += premium_ticket_price
        count += 1
        if count == aircraft_capacity:
            break
    print("Number of Prem tickets:", abs(count- int(standard_ticket_demand)))
    print("total tickets:", count)    
    print("revenue from prem tickets :", revenue_premium)
    
    total_revenue = revenue_standard + revenue_premium
    tot_revenue.append(total_revenue)
    print("total rev:", tot_revenue)  
    print("------------------")

    
#print(tot_revenue)  
avg_tot_revenue = np.mean(tot_revenue)
print("Avg Rev:",avg_tot_revenue)
```
