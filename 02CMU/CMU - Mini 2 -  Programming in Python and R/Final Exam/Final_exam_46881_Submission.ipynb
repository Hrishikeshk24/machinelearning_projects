### PROBLEM 2 (45 points)

In this problem I ask you to implement in Python a simulation model of a simpified airline revenue management problem.

Consider an airline operating a flight with an aircraft whose capacity is 189 passenger seats. All seats on this aircraft are identical (that is, there is no business or first-class compartments). The marginal cost of each additional passenger is negligible. 

There are two fares for this flights: Standard and Premium. Standard tickets cost 249 dollars and have advance-purchase and other restrictions. Premium tickets cost 349 dollars and have far fewer restrictions.

Because of the advance-purchase restriction of the Standard tickets and the flexibility of the Premium tickets we will assume that **all** of the demand for Standard tickets materializes before any of the demand for Premium tickets. In other words, Standard fare passengers book early and Premium fare passengers book later.

It is estimated that the demand for Standard tickets follows approximately a Normal distribution with mean equal to 150 and a standard deviation equal to 30. The demand for Premium tickets is estimated to also follow approximately a Normal distribution but with mean equal to 100 and a standard deviation equal to 40. Assume the demands to be uncorrelated. (Do not worry about fractionality or the possibility of negative demand values).

For each part of this problem use 10,000 trials per simulation and a value of 123 for the random seed.

#### Part (a) 

Suppose the airline accepts all bookings, subject to capacity, on a first-come first-serve basis. Run a simulation to compute the expected (or average) revenue per flight?

### Approach: 

* In part (a), as there is no limit set on the number of standard tickets that can be purchased, we will use the random number generator normal function to determine the demand for the standard fare. The cap of 189 is applied to make sure airline accepts all bookings, subject to aircraft capacity. Using the determined standard fare demand, we calculate the 'standard revenue'.

* Next, as we assume that all of the demand for Standard tickets materializes before any of the demand for Premium tickets, based on the remaining number of seats in the aircraft after allocating standard fare seats, we run the simulation for the premium demand until the aircraft capacity is reached and calculate the corresponding 'premium revenue'.

* Total revenue is a summation of 'standard revenue' and 'premium revenue'.

* After simulating the above for 10000 trials, we take the average to determine the average revenue per flight.

```python
# Importing required libraries/packages
import numpy as np
```

```python
# Defining required variables

aircraft_capacity = 189
standard_ticket_price = 249
premium_ticket_price = 349
std_ticket_mean = 150
std_ticket_stdev = 30
premium_ticket_mean = 100
premium_ticket_stdev = 40
trials = 10000
```

```python
tot_revenue =[]
rng = np.random.default_rng(seed=123)
for i in range(trials):
    
    standard_ticket_demand = rng.normal(loc= std_ticket_mean, scale = std_ticket_stdev)
    if standard_ticket_demand >= aircraft_capacity:
        total_seats_filled = aircraft_capacity
    else:
        total_seats_filled = int(standard_ticket_demand)
       
    revenue_standard = total_seats_filled * standard_ticket_price
       
    revenue_premium = 0
    if total_seats_filled < aircraft_capacity:
        premium_ticket_demand = rng.normal(loc = premium_ticket_mean, scale = premium_ticket_stdev)
        for demand in range(int(premium_ticket_demand)):
            revenue_premium += premium_ticket_price
            total_seats_filled += 1
            if total_seats_filled == aircraft_capacity:
                break
        total_revenue = revenue_standard + revenue_premium
        tot_revenue.append(total_revenue)
    else:
        total_revenue = revenue_standard + revenue_premium
        tot_revenue.append(total_revenue)

avg_tot_revenue = np.mean(tot_revenue)
print("The average revenue per flight is:", avg_tot_revenue)

```

#### Part (b) 

Accepting customers on a first-come first-serve effectively dilutes revenue. Passengers requesting Standard tickets books early and this limits (due to aircraft capacity) the airline's ability to issue higher-fare Premium tickets for passegers booking later. Therefore, the airline's revenue management system will implement a **booking limit** on Standard tickets. That is, it will set a limit on the maximum number of Standard tickets that will be issued on any given flight (even before it knows precisely the demand for Premium tickets). Requests for Standard ticket bookings declined due to the booking limit is lost demand.  

Run a simulation to compute the expected (or average) revenue per flight if the airline chooses to set a booking limit of 110 seats.

### Approach:

* The approach is similar to part(a) except for the fact that the booking limit of 110 is set on the max number of tickets that can be purchased using the standard fare per flight.

```python
booking_limit = 110
```

```python
tot_revenue =[]
rng = np.random.default_rng(seed=123)
for i in range(trials):
    
    standard_ticket_demand = rng.normal(loc= std_ticket_mean, scale = std_ticket_stdev)
    if standard_ticket_demand >= booking_limit:
        total_seats_filled = booking_limit
    else:
        total_seats_filled = int(standard_ticket_demand)
       
    revenue_standard = total_seats_filled * standard_ticket_price 
    
    revenue_premium = 0
    premium_ticket_demand = rng.normal(loc = premium_ticket_mean, scale = premium_ticket_stdev)
    for demand in range(int(premium_ticket_demand)):
        revenue_premium += premium_ticket_price
        total_seats_filled += 1
        if total_seats_filled == aircraft_capacity:
            break
    total_revenue = revenue_standard + revenue_premium
    tot_revenue.append(total_revenue)

    
avg_tot_revenue = np.mean(tot_revenue)
print("The average revenue per flight is:", avg_tot_revenue)
```

### Conclusion:
