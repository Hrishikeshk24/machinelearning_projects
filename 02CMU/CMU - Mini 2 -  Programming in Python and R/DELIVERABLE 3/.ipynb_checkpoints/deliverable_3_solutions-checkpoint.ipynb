### Setting up

```python
import numpy as np
import pandas as pd
import copy
```

### Importing and Exploring Data 

```python
consultants = pd.read_csv("consultants.csv").to_dict("records")
```

```python
num_consultants = len(consultants)
```

```python
num_consultants
```

```python
num_teams = 20
```

```python
num_analytics_expertise = sum(c["analytics_expertise"] for c in consultants)
num_functional_expertise = sum(c["functional_expertise"] for c in consultants)
num_domain_expertise = sum(c["domain_expertise"] for c in consultants)
num_communication_skill = sum(c["communication_skill"] for c in consultants)
print("Number of consultants with Analytics expertise:", num_analytics_expertise)
print("Number of consultants with Functional expertise:", num_functional_expertise)
print("Number of consultants with Domain expertise:", num_domain_expertise)
print("Number of consultants with Communication skills:", num_communication_skill)
```

### Basic Algorithm

**The basic idea is to sort the 100 consultants according to some criterion and then assign that sequence to teams in a rotating fashion: 1,...,20,1,...20,1,...,20,1,...,20,1,...,20. This idea is implemented in the function below. The team assignment is denoted by a new key added to each consultant's dictionary**.

```python
def rotating_allocation(consultant_sequence):
    rotating_teams = list(range(1,num_teams+1)) * 5
    for i in range(len(consultant_sequence)):
        consultant_sequence[i]["team"] = rotating_teams[i]
    return(consultant_sequence)    
```

### Functions that summarize and print the resulting allocation

```python
def summarize(allocation):
    summary = list()
    for t in range(1,num_teams+1):
        num_a = sum(allocation[i]["analytics_expertise"] for i in range(num_consultants) if allocation[i]["team"] == t)
        num_f = sum(allocation[i]["functional_expertise"] for i in range(num_consultants) if allocation[i]["team"] == t)
        num_d = sum(allocation[i]["domain_expertise"] for i in range(num_consultants) if allocation[i]["team"] == t)
        num_c = sum(allocation[i]["communication_skill"] for i in range(num_consultants) if allocation[i]["team"] == t)
        team_summary = {"Team": t, "Analytics": num_a, "Functional": num_f, "Domain": num_d, "Communication": num_c}
        summary.append(team_summary)
    return(summary)    
```

```python
def print_summary(summary):
    for t in summary:
        print("Team", "{:2}".format(t["Team"]),":",
              "Analytics = ", t["Analytics"],
              "Functional = ", t["Functional"],
              "Domain = ", t["Domain"],
              "Communication = ", t["Communication"])
```

### Approach 0 - Original sequence

```python
sequence_0 = copy.deepcopy(consultants) 
```

```python
allocation_0 = rotating_allocation(sequence_0)
```

```python
print_summary(summarize(allocation_0))
```

### Approach 1 - Sort by sum of skills

```python
sequence_1 = copy.deepcopy(consultants) 
```

```python
sequence_1 = sorted(sequence_1, 
                    key=lambda c: c["analytics_expertise"]+c["functional_expertise"]+c["domain_expertise"]+c["communication_skill"],
                    reverse=True) 
```

```python
allocation_1 = rotating_allocation(sequence_1)
```

```python
print_summary(summarize(allocation_1))
```

### Approach 2 - Sort in order of rarest expertise first

```python
sequence_2 = copy.deepcopy(consultants) 
```

```python
sequence_2 = sorted(sequence_2, 
                    key=lambda c: (c["analytics_expertise"],c["domain_expertise"],c["functional_expertise"],c["communication_skill"]),
                    reverse=True) 
```

```python
allocation_2 = rotating_allocation(sequence_2)
```

```python
print_summary(summarize(allocation_2))
```

### Approach 3 - Sort in order of most common expertise first

```python
sequence_3 = copy.deepcopy(consultants) 
```

```python
sequence_3 = sorted(sequence_3, 
                    key=lambda c: (c["communication_skill"],c["functional_expertise"],c["domain_expertise"],c["analytics_expertise"]),
                    reverse=True) 
```
