## Python pandas: Column-wise operations

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("PIT_2021.csv")
```

```python
df.columns
```

#### Deleting (dropping) columns

```python
df = df.drop(["YEAR"],axis="columns") # or axis = 1
```

```python
df.columns
```

#### Adding a column with specified values

```python
df["YEAR"] = 2021
```

```python
df.columns
```

#### Re-arranging (Re-indexing) columns

```python
col_names = list(df.columns)
```

```python
col_names.insert(0,col_names.pop())
```

```python
col_names
```

```python
df = df.reindex(columns = col_names)
```

```python
df
```

#### Adding computed columns

```python
df["DEP_DELAY_HRS"] = df["DEP_DELAY"] / 60
```

```python
df
```

#### Renaming columns

```python
df = df.rename(columns={"DEP_DELAY_HRS": "departure delay in hours"})
```

```python
df
```
