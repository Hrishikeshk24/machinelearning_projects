## Python pandas: Data structures

```python
import numpy as np
import pandas as pd
```

#### The pandas Series

A pandas Series is a **one-dimensional** array(-like) object consisting of values **of the same data type** along with an associated array of (row) labels called the **index**. 

Suppose we have two stores generating revenue in Q1. We can create a Series from a python list using the Series function. For example:

```python
Q1_revenue = pd.Series([1000, 2500])
print(Q1_revenue)
```

```python
Q1_revenue.index
```

We can specify a custom index:

```python
Q1_revenue = pd.Series([1000, 2500],
                       index=["store A", "store B"])
print(Q1_revenue)
```

Series elements are accessible by their indices:

```python
print(Q1_revenue["store B"])
```

Indices are preserved during arithmetic operations: 

```python
Q1_revenue_after_tax = 0.7 * Q1_revenue
print(Q1_revenue_after_tax)
```

Indices are also preserved during filtering operations: 

```python
Q1_revenue[Q1_revenue > 1000]
```

```python
Q1_revenue
```

We can also create a pandas Series from a Python dictionary using the Series function. For example: 

```python
Q2_revenue = pd.Series({"store C": 500, "store A": 1300, "store B": 2900})
print(Q2_revenue)
```

Indices align data rows properly. For example:

```python
cumulative_revenue = Q1_revenue + Q2_revenue
print(cumulative_revenue)
```

The index object can be retrieved using the index attribute of a series:

```python
print(list(cumulative_revenue.index))
```

```python
cumulative_revenue.index
```

#### The pandas DataFrame

A pandas DataFrame is a **rectangular** data structure consisting of multiple columns. Each column has a specific data type, but **the data types can vary across columns**. 

You can think of a pandas DataFrame as a dictionary of pandas Series, all **sharing the same index**.

```python
Q1_revenue
```

```python
Q2_revenue
```

```python
revenue_df = pd.DataFrame({"Q1": Q1_revenue, "Q2": Q2_revenue})
print(revenue_df)
```

You can create a DataFrame directly from a dictionary of equal-length lists.

```python
Q2_df = pd.DataFrame({"revenue": [500, 1300, 2900],
                      "costs": [200, 500, 700]},
                      index = ["store C", "store A", "store B"])
print(Q2_df)
```

#### Retrieving row indices and column names

```python
print(Q2_df.index)
```

```python
print(list(Q2_df.columns))
```

#### Browser display

```python
Q2_df
```

#### Pointers vs. Copies

```python
df1 = Q2_df
```

```python
df1
```

```python
df1["profit"] = df1["revenue"] - df1["costs"]
```

```python
df1
```

```python
Q2_df
```

*The assignment (=) copies the pointer to the location of the dataframe. It does not create a separate copy the dataframe.* 

What if we want a separate copy of the dataframe?

```python
Q2_df = pd.DataFrame({"revenue": [1300, 2900, 500],
                      "costs": [500, 700, 200]},
                      index = ["store A", "store B", "store C"])
```

```python
Q2_df
```

```python
df2 = Q2_df.copy()
```

```python
df2
```

```python
df2["profit"] = df2["revenue"] - df2["costs"]
```

```python
df2
```

```python
Q2_df
```

*Use the copy method to create a separate copy the dataframe.*
