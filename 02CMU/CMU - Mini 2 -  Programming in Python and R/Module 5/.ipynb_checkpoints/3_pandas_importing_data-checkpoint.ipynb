## Python pandas: Importing data

```python
import numpy as np
import pandas as pd
```

#### Introducing the PIT departures dataset

Place the accompanying dataset "PIT_2021.csv" **in the same folder location as this Notebook**. The dataset has flight delay information pertainig to all departures out of PIT (Pittsburgh International Airport) in 2021. 

Data source: Bureau of Transportation Statistics.
Covers domestic flights only
https://www.transtats.bts.gov/DL_SelectFields.aspx?gnoyr_VQ=FGJ&QO_fu146_anzr=b0-gvzr

The dataset contains scheduled and actual departure and arrival times reported by certified U.S. air carriers that account for at least one percent of domestic scheduled passenger revenues. The data is collected by the Office of Airline Information, Bureau of Transportation Statistics (BTS).

Accompanying reference files "Documentation.xlsx" and "Carrier_codes.xlsx" are available on Canvas.

#### Importing csv data as a DataFrame

```python
df = pd.read_csv("PIT_2021.csv")
```
