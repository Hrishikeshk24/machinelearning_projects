## Python pandas: Selection in DataFrames

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"revenue": [500, 1300, 2900],
                    "costs": [200, 500, 700],
                   "profit": [300, 800, 2200]},
                    index = ["store A", "store B", "store C"])
```

```python
df
```

#### Selecting a single value by row and column label

```python
df.at["store B", "costs"]
```

#### Selecting a single value by row and column integer position

```python
df.iat[2, 1]
```

#### Selecting both row(s) and column(s) by label

```python
df.loc[["store B", "store C"], ["revenue", "costs"]]
```

#### Selecting both row(s) and column(s) by integer position

```python
df.iloc[[2],[0]]
```

#### Selecting row(s) by label

```python
df.loc[["store B", "store C"]]
```

#### Selecting row(s) by integer position

```python
df
```

```python
df.iloc[[2,0]]
```

#### Selecting columns(s) by label

```python
df.loc[:, ["revenue","costs"]]
```

```python
df[["revenue","costs"]]
```

#### Selecting columns(s) by integer position

```python
df.iloc[:, [0,1]]
```
