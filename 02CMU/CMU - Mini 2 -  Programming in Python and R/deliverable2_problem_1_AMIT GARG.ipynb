```python
import numpy as np
import matplotlib.pyplot as plt
Beta = 0.10
Gamma = 0.05
days = 365
N = 1000
S = np.zeros (days)
I = np.zeros (days)
R = np.zeros (days)
S [0] = 990
I [0] = 10
R [0]= 0

for t in range(days-1):
    S[t+1] = (S[t]-((Beta*S[t]*I[t])/N))
    I[t+1] = (I[t]+((Beta*S[t]*I[t])/N)- Gamma * I[t])
    R[t+1] = (R[t]+ Gamma * I[t])
    
#print(I)
plt.grid()
plt.plot(S, label = "Susceptible")
plt.plot(I, label = "Infections")
plt.plot(R, label = "Recovered")
plt.xlabel('days')
plt.ylabel('Numbers of individuals')
plt.legend()
plt.show()
```

Suppose our healthcare system has a capacity to accommodate a maximum of 120 infectious persons. Is this capacity sufficient according to the model?

Ans. 
The maximum number of individuals that will be infected is 160. So, this capacity is not sufficient according to model.

Part (b)
Now suppose that our healthcare system can currently accommodate only 120 infectious persons. However, its capacity can be increased to accommodate 180 infectious persons in 2 months. Is this current capacity and expansion plan sufficient according to the model?

Ans.
The maximum number of individuals that will be infected in 2 months (61 days) is 112. The current healthcare system can currenlty accomodate 120 infectious persons. So, current capacity is sufficient.

The maximum number of individuals that will be infected is 160 (in ~ 3 months). The healthcare system capciaty will increase to accomodate 180 infectious persons in 2 months. So, this expansion plan is sufficient.
