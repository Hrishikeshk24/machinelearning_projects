#### Example 1: Defining a standard deviation function

```python
def my_sd_fun(list_of_numbers):
    av = np.mean(list_of_numbers)
    sum_squares = 0
    for x in list_of_numbers:
        sum_squares = sum_squares + (x - av)**2
    variance = sum_squares / len(list_of_numbers)    
    return np.sqrt(variance)    
```

```python
a_list = [2, 4, 3, -1]
```

```python
my_sd_fun(a_list)
```

```python
# compare with numpy's built-in std function
np.std(a_list)
```

#### Example 2: Implementing a profit function in the order quantity application

```python
unit_price = 10
unit_cost = 5
unit_salvage_value = 3
```

Defining a profit function:

```python
def compute_profit(q, d):
    total_cost = unit_cost * q 
    if d <= q:
        units_sold = d
        units_salvaged = q - d
    else:  
        units_sold = q
        units_salvaged = 0 
        
    sales_revenue = unit_price * units_sold
    salvage_revenue = unit_salvage_value * units_salvaged
    total_revenue = sales_revenue + salvage_revenue

    profit = total_revenue - total_cost
    return profit
```

Let's assume that demand follows a *Binomial distribution* with the following parameters:

```python
n = 400
p = 0.25
```

```python
order_quantity_list = list(range(80,121,1))
expected_profit_list = []
for order_quantity in order_quantity_list:
    profit_list = []
    rng = np.random.default_rng(seed=881)
    for i in range(10000):
        demand = rng.binomial(n, p)
        profit = compute_profit(order_quantity, demand)
        profit_list.append(profit)
    av_profit_for_given_quantity = np.mean(profit_list)
    expected_profit_list.append(av_profit_for_given_quantity)
print(expected_profit_list)    
```

```python
max_profit = np.max(expected_profit_list)
optimal_quantity = order_quantity_list[expected_profit_list.index(max_profit)]
print("Ordering", optimal_quantity, "units yields the optimal profit of", max_profit, "dollars.")
```
