## Python dictionaries

#### Defining a dictionary

```python
cust1 = {'balance': 15000, 'income': 100000, 'credit_score': 750}
```

```python
type(cust1)
```

```python
cust2 = {'balance': 8000, 'income': 200000, 'credit_score': 780}
cust3 = {'balance': 10000, 'income': 70000, 'credit_score': 800}
```

```python
# A dictionary of dictionaries
my_customers = {'id1': cust1, 'id2': cust2, 'id3': cust3}
```

#### Getting values from a dictionary

```python
cust1['credit_score']
```

```python
my_customers['id1']['credit_score']
```

#### keys method

```python
my_customers.keys()
```

```python
list(my_customers.keys())
```

```python
total_balance = 0
for cust in my_customers.keys():
    total_balance = total_balance + my_customers[cust]['balance']
print(total_balance)    
```
