## More on Python lists

#### Defining a list

```python
# Defining a list of values
week_days = ["Mon", "Tue", "Wed", "Thr", "Fri", "Sat", "Sun"]
```

```python
# Checking the variable type
type(week_days)
```

```python
# Defining an empty list
empt_list = []
```

#### Getting values from a list

```python
# Accessing indivdual elements of a list
print(week_days[0])
```

```python
# Accessing indivdual elements from the end of a list 
print(week_days[-2])
```

```python
# List slicing (elements 1 and 2)
print(week_days[1:3])
```

```python
# List slicing (elements 1 through the end)
print(week_days[1:])
```

```python
# List slicing (elements from the start through to but not including the end index)
print(week_days[:3])
```

#### Returning the length of a list

```python
len(week_days)
```

#### Deleting a list element by index

```python
del week_days[-1]
print(week_days)
```

#### Deleting a list element by value

```python
week_days.remove('Fri')
print(week_days)                 
```

#### Adding list elements

```python
week_days.append('Sun')
print(week_days) 
```

```python
week_days.insert(4,'Fri')
print(week_days) 
```

#### Sorting a list

Uses "ASCIIbetical order"...

```python
week_days.sort()
print(week_days)
```

```python
week_days.sort(reverse=True)
print(week_days)
```

#### Getting the index of a value

```python
week_days.index('Thr')
```

#### Changing the value of list elements

```python
Thr_index = week_days.index('Thr')
print(Thr_index)
```

```python
week_days[Thr_index] = 'Thu'
print(week_days)
```

#### Lists work well in "for loops" 

```python
for day in week_days:
    print(day)
```
