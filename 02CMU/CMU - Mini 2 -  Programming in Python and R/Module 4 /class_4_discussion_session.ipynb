### 1.3. User-defined functions

#### Example: Defining a standard deviation function

```python
def my_sd_fun(list_of_numbers):
    av = np.mean(list_of_numbers)
    sum_squares = 0
    for x in list_of_numbers:
        sum_squares = sum_squares + (x - av)**2
    variance = sum_squares / len(list_of_numbers)    
    return np.sqrt(variance)    
```

```python
a_list = [2, 4, 3, -1]
```

```python
my_sd_fun(a_list)
```

```python
# compare with numpy's built-in std function
np.std(a_list)
```

#### Defining a jobs sequence using a list

```python
default_sequence = [j1,j2,j3,j4,j5,j6,j7,j8,j9,j10,j11,j12,j13,j14,j15,j16,j17,j18,j19,j20]
```

#### Processing the schedule

```python
def compute_schedule(sequence):
    schedule = copy.deepcopy(sequence) 
    t = 0
    for job in schedule:
        proc = job["processing_time"]
        job["start"] = t
        job["complete"] = t + proc
        job["lateness"] = max(0, job["complete"] - job["due_date"])
        if job["lateness"] > 0:
            job["late"] = 1
        else:
            job["late"] = 0
        t = t + proc
    return(schedule)
```

```python
schedule = compute_schedule(default_sequence)
print(schedule)
```

```python
count =0 
for i in range(len(compute_schedule(default_sequence))):
    #print(i)
    #print(compute_schedule(default_sequence)[i]['late'])
    if compute_schedule(default_sequence)[i]['late'] == 1:
        count += 1
        
print(count)
    
```

#### Computing performance metrics

```python
def compute_num_jobs_late(schedule):
    num_jobs_late = sum(job["late"] for job in schedule)
    return num_jobs_late   
```

```python
def compute_max_lateness(schedule):
    max_lateness = max(job["lateness"] for job in schedule)
    return max_lateness  
```

```python
def compute_sum_completion_times(schedule):
    sum_completion_times = sum(job["complete"] for job in schedule) 
    return sum_completion_times
```

#### Evaluating the default sequence

```python
default_schedule = compute_schedule(default_sequence)
```

```python
num_jobs_late = compute_num_jobs_late(default_schedule)
print("Using the default sequence, the number of jobs late is ", num_jobs_late)
```

```python
max_lateness = compute_max_lateness(default_schedule)
print("Using the default sequence, the maximum lateness is ", max_lateness)
```

```python
sum_completion_times = compute_sum_completion_times(default_schedule)
print("Using the default sequence, the sum of completion times ", sum_completion_times)
```

Q1) Can we write a function that **minimizes max_lateness**?

```python
def seq_earliest_due_date(jobs_list):
    sequence = sorted(jobs_list, key=lambda j: j['due_date']) 
    return(sequence) 
```

```python
earliest_due_date_sequence = seq_earliest_due_date(default_sequence)
earliest_due_date_schedule = compute_schedule(earliest_due_date_sequence)
```

```python
minimum_max_lateness = compute_max_lateness(earliest_due_date_schedule)
print("Using the earliest due date sequence, the maximum lateness is ", minimum_max_lateness)
```

Q2) Can we write a function that **minimizes sum_completion_times**?

```python
def seq_shortest_processing_time(jobs_list):
    sequence = sorted(jobs_list, key=lambda d: d['processing_time']) 
    return(sequence)
```

```python
shortest_proc_time_sequence = seq_shortest_processing_time(default_sequence)
shortest_proc_time_schedule = compute_schedule(shortest_proc_time_sequence)
```

```python
min_sum_completion_times = compute_sum_completion_times(shortest_proc_time_schedule)
print("Using the shortest processing time rule, sum of completion times is ", min_sum_completion_times)
```
