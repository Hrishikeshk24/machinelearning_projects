### Section 2: Exploring Departure Time Patterns

**Task 2.1**: Compute the number of flights departing each month. What do you observe? Hint: The simplest way is to use the "value_counts()" on the MONTH series; optionally, you may then choose to apply the  ".sort_index()" method for readability.

```python
df.pivot_table(index=["MONTH"],aggfunc= "count", values=["OP_CARRIER"])
```

```python
df["MONTH"].value_counts().sort_index()
```

**Task 2.2**: Compute the number of flights departing each day of the week (Monday through Sunday). What do you observe? Note (a): Monday=1, Sunday=7; Note (b): 2021 started on a Friday and ended on a Friday (so one more Friday in the year than the other week days). On average, which day(s) of the week are busiest in terms of departures.

```python
df.pivot_table(index=["DAY_OF_WEEK"],aggfunc= "count", values=["OP_CARRIER"])
```

```python
df["DAY_OF_WEEK"].value_counts().sort_index()
```

**Task 2.3**: Compute the number of flights scheduled to depart each hour of the day. Hint: Use "DEP_TIME_BLK". What do you observe? 

```python
df.pivot_table(index=["DEP_TIME_BLK"],aggfunc= "count", values=["OP_CARRIER"])
```

**Task 2.4**: Compute the fraction of flights whose departure is delayed 15 or more mins for each hour of the day. What do you observe? Hint: Use the groupby() and mean() methods.

```python
df.pivot_table(index=["DEP_TIME_BLK"],aggfunc= "mean", values=["DEP_DEL15"])
```

### Section 4: A More Detailed Look at Airline Delay / On-Time Statistics

**Task 4.1**: Create two new columns in "df": "DEP_ON_TIME" and "ARR_ON_TIME". A flight is on time if it's not cancelled or diverted or past 15 (actually 14) minutes past schedule. Hint: You can create a logical expression and then convert, if you wish, to 0 (False) and 1 (True) using the ".astype(int)" method. You may want to inspect the resulting dataframe after adding the columns using head(), tail(), and/or .info. 

```python
df["ARR_ON_TIME"] = ((df["ARR_DEL15"] == 0) & (df["CANCELLED"] == 0) & (df["DIVERTED"] == 0)).astype(int)
```

```python
df["DEP_ON_TIME"] = ((df["DEP_DEL15"] == 0) & (df["CANCELLED"] == 0) & (df["DIVERTED"] == 0)).astype(int)
```

```python
df
```

**Task 4.2**: Now create the following list of delay / on-time performance-related columns.

**Task 4.3**: Create a table comparing all airlines in terms of these performance metrics (using the average of the columns). Sort the table based on "DEP_DEL15" as you've done in Task 2.3. Pause and examine carefully the results. 

```python
perf_cols = ["ARR_DELAY", "ARR_DELAY_NEW", "ARR_DEL15", "ARR_ON_TIME"]
```

```python
df.pivot_table(index ='OP_CARRIER', aggfunc = 'mean', values = perf_cols).sort_values(by=['ARR_ON_TIME'])
```
