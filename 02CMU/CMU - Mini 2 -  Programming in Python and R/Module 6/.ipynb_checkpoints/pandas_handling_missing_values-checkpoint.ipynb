## Handling NA values

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("covid19_US_V.csv", )
```

### Identifying NA values

```python
df
```

```python
df.info()
```

```python
df.isna()
```

```python
df.isna()["vaccination_rate"]
```

```python
df.isna()["vaccination_rate"].sum()
```

```python
df.notna()["vaccination_rate"].sum()
```

```python
#df.isna().any(axis=1)
df[df.isna().any(axis=1) == True]
```

### Note: pandas by default ignores NA values

```python
df["vaccination_rate"].median()
```

```python
df["vaccination_rate"].median(skipna=False)
```

### Handling NA values

#### Option 0: Do nothing (and interpret summary stats / model fit accordingly)

#### Option 1: Filter out the NA values

By dropping rows where any of the values is NaN.

```python
df.dropna()
```

Or, by dropping only rows where *all* of the values are NaN.

```python
df.dropna(how="all")
```

#### Option 2 (advanced): Impute the NA values in numerical columns

Usually requires some modeling...
The simplest case is to replace with the average (or median value)

```python
df["vaccination_rate"].fillna(df["vaccination_rate"].mean())
```
