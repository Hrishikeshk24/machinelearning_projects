## Pandas Pivot Tables

Also, see Chapter 10 of https://wesmckinney.com/book/.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("PIT_rideshare_dataset.csv")
```

```python
df.columns
```

```python
df.pivot_table(index="month", aggfunc="mean")
```

```python
df.pivot_table(index="month", aggfunc = "count")
```

```python
df.pivot_table(index="month", aggfunc = "count", values=["rentals","temperature"])
```

```python
df.pivot_table(index=["month", "day_of_week"], aggfunc = "mean", values=["rentals","temperature"])
```

```python
df.pivot_table(index=["month"], columns=["day_of_week"], aggfunc = "mean", values=["rentals"])
```

```python
df.pivot_table(index=["month"], columns=["day_of_week"], aggfunc = "mean", values=["rentals"], margins=True)
```
