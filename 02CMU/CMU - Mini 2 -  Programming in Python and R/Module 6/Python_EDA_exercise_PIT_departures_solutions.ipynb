### Section 4: A More Detailed Look at Airline Delay / On-Time Statistics

**Task 4.1**: Create a new columns in "df": "ARR_ON_TIME". A flight is on time if it's not cancelled/diverted *and* arrives earlier than 15 minutes past its scheduled arrival time. 

```python
df["ARR_ON_TIME"] = ((df["ARR_DEL15"] == 0) & (df["CANCELLED"] == 0) & (df["DIVERTED"] == 0)).astype(int)
```

**Task 4.2**: Create a table comparing all airlines in terms of these performance metrics (using the average of the columns). Which airline(s) have the best on-time performance?

```python
perf_cols = ["ARR_DELAY", "ARR_DELAY_NEW", "ARR_DEL15", "ARR_ON_TIME"]
```

```python
#df[perf_cols].groupby(df["OP_CARRIER"]).mean().sort_values("DEP_DEL15", ascending=False)
df.pivot_table(index="OP_CARRIER", values=perf_cols, aggfunc="mean").sort_values(by="ARR_ON_TIME", ascending=False)
```

*Delta's performance beats any other airline across all metrics*

*Compared to American Airlines, Southwest is doing worse on ARR_DEL15 and ARR_ON_TIME but better on ARR_DELAY_NEW (minutes) which suggests that when they're delayed they're not delayed by much.*

**Task 4.3**: Examine if these statistics are comparable by looking at patterns *where* airlines fly.

```python
#pd.crosstab(df["OP_CARRIER"], df["DEST_STATE_ABR"], margins=True, normalize="index")
pd.options.display.max_rows = 20
#df.pivot_table(index=["OP_CARRIER","DEST_STATE_ABR"], values="ARR_ON_TIME", aggfunc="mean")
df.pivot_table(index=["OP_CARRIER"],columns=["DEST_STATE_ABR"], values="ARR_ON_TIME", aggfunc="mean")
```

*The airlines serve different destinations. It could be that congestion at destination (possibly also affecting incoming flight) is responsible for the delays. We cannot exclude Simpson's type paradoxes when comparing these airlines on aggregate metrics.*
