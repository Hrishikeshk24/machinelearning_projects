## Combining DataFrames

```python
import numpy as np
import pandas as pd
```

#### Creating small dataframes for illustrative purposes

```python
df_pop1 = pd.DataFrame({"state": ["Pennsylvania", "Ohio"],
                       "population": [12801989, 11689100]})
df_pop1
```

```python
df_pop2 = pd.DataFrame({"state": ["New Jersey", "New York"],
                        "population": [8882190, 19453561]})
df_pop2
```

```python
df_vax = pd.DataFrame({"state": ["Pennsylvania", "Ohio", "New York", "West Virginia"],
                       "people_fully_vaccinated": [9117526, 6969367, 15391715, 1056958]})
df_vax
```

#### Binding (concatenating) rows or columns of two dataframes

```python
df_pop = pd.concat([df_pop1,df_pop2])
```

```python
df_pop
```

#### Joining two dataframes 

```python
pd.merge(df_pop, df_vax, on="state")
```

```python
pd.merge(df_pop, df_vax, on="state", how="inner") ## inner is default. See above
```

By default, pandas.merge does an "inner" join

Types of merge (joins): inner, left, right, outer

```python
pd.merge(df_pop, df_vax, on="state", how="left")
```

```python
pd.merge(df_pop, df_vax, on="state", how="right")
```

```python
pd.merge(df_pop, df_vax, on="state", how="outer")
```
