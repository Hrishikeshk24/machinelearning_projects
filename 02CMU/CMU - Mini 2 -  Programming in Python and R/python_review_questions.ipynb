6.1. Define a function that sums the squares of the elements of any list. Test it on this list [0, 1, 1, 2, 3]

```python
def sum_of_squares(list):
    new_list=[]
    for i in list:
        new_list.append(i**2)
    return sum(new_list)
```

```python
sum_of_squares([0, 1, 1, 2, 3])
```

```python
def fun2(list_input):
    list_squared = [x**2 for x in list_input]
    sum_squares = sum(list_squared)
    return(sum_squares)
```

```python
fun2([0, 1, 1, 2, 3])
```

6.2. Define a function that takes two equal-length lists, a list of predicted numerical values and a list of actual numerical values, and calculates the root mean square error (or deviation) between these two lists. For example, if pred = [1.5, 2.0, 2.5] and actual = [1, 2, 3]  then the root means square error is: 
$$\sqrt{\frac{(1.5-1)^{2}+(2.0-2)^{2}+(2.5-3)^{2}}{3}} = 0.408 $$

```python
def rmse(pred, actual):
    error = np.subtract(pred, actual)
    mean_square_error = fun2(error) / len(error)
    return(np.sqrt(mean_square_error))
```

```python
# Test
rmse([1.5, 2.0, 2.5], [1, 2, 3])
```
