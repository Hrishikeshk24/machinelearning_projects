## Conditionals: The "if else" statement

Checkout the basic sytax of an if else statement here:
https://docs.python.org/3/tutorial/controlflow.html#if-statements

**Example: The order quantity problem**

```python
unit_price = 10
unit_cost = 5
unit_salvage_value = 3
```

```python
n = 400
p = 0.25
```

Let's start with a very simple calculation. Suppose we order 100 units (the expected demand) but demand turns out to equal 80 units. What is the profit in this case?

```python
order_quantity = 100
```

```python
demand = 80
```

Calculating profit:

```python
total_cost = unit_cost * order_quantity 

if demand <= order_quantity:
    units_sold = demand
    units_salvaged = order_quantity - demand
else:  
    units_sold = order_quantity
    units_salvaged = 0
    
sales_revenue = unit_price * units_sold
salvage_revenue = unit_salvage_value * units_salvaged
total_revenue = sales_revenue + salvage_revenue

profit = total_revenue - total_cost
print(profit)
```
