```python
import pandas as pd
import numpy as np
```

```python
df_movies = pd.read_csv("movies.csv")
```

```python
type(df_movies)
```

```python
df_movies
```

```python
df_ratings = pd.read_csv("ratings.csv")
```

```python
type(df_ratings)
```

```python
df_ratings
```

```python
df_combined = pd.merge(df_movies, df_ratings, on='movieId').sort_values(by=['movieId'])
```

```python
df_combined
```

```python
df_combined.tail(5)
```

```python
mean_ratings = df_combined.pivot_table(index = 'title', aggfunc ='mean',values = 'rating').sort_values(by='rating')
```

```python
mean_ratings.head(5)
```

```python
ratings_by_title = df_combined.groupby("title").size()
```
