#### Part (h)
Create a new column in df called "Delay indicator". Set this column to equal 1 if "Arrival delay minutes" equals or exceeds 15 minutes, and zero otherwise. If a flight has NaN in the "Arrival delay minutes" column then "Delay indicator" should display NaN as well. (Note: The FAA defines a flight as "delayed" if it arrives at its destination 15 or more minutes past its scheduled arrival time).

```python
# create a list of conditions
conditions = [
    (df['Arrival delay minutes'] < 15),
    (df['Arrival delay minutes'] >= 15),
    (pd.isnull(df['Arrival delay minutes']))
    ]
```

```python
# create a list of the values
values = [0, 1, "NaN"]
```

```python
# create a new column -  Delay indicator and assign values to it

df['Delay indicator'] = np.select(conditions, values)
```

```python
df
```

#### Part (i)
What is the percentage of delayed flights (as indicated by the "Delay indicator" column) for Optima Airlines?

```python
df_Optima = df[df["Airline"] == "Optima"]
```

```python
total_O = len(df_Optima) - len(df_Optima[df_Optima['Delay indicator']== 'NaN']) 
delayed_O = len(df_Optima[df_Optima['Delay indicator']== '1'])
print(total_O)
print(delayed_O)

perc_delayed_Opt_flights = delayed_O / total_O

print(perc_delayed_Opt_flights * 100, '%')
```

#### Part (j)
What is the percentage of delayed flights (as indicated by the "Delay indicator" column) for Quanta Airlines?

```python
df_Quanta = df[df["Airline"] == "Quanta"]
```

```python
total_Q = len(df_Quanta) - len(df_Quanta[df_Quanta['Delay indicator']== 'NaN']) 
delayed_Q = len(df_Quanta[df_Quanta['Delay indicator']== '1'])
print(total_Q)
print(delayed_Q)

perc_delayed_Quanta_flights = delayed_Q / total_Q

print(perc_delayed_Quanta_flights * 100, '%')
```
