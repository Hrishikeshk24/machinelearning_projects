```python
categori=['job', 'marital', 'education', 'default','housing', 'loan', 'contact', 'month', 'day', 'poutcome','y']
for col in categori:
    plt.figure(figsize=(11,6))
    sns.barplot(data[col].value_counts(),data[col].value_counts().index,data=data)
    plt.title(col)
    plt.tight_layout()
```
