### Create a baseline model

Some utility functions.

Note that we use stratified cross-validation to ensure that the distribution
of class labels in each fold is the same as the distribution of the class
labels in the full dataset.

```python
# load the dataset
def load_dataset(full_path):
    # load the dataset as a numpy array
    data = read_csv(full_path, header=None)
    # retrieve numpy array
    data = data.values
    # split into input and output elements
    X, y = data[:, :-1], data[:, -1]
    # label encode the target variable to have the classes 0 and 1
    y = LabelEncoder().fit_transform(y)
    return X, y


# evaluate a model
def evaluate_model(X, y, model):
    # define evaluation procedure
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    # evaluate model
    scores = cross_val_score(model, X, y, scoring="f1", cv=cv, n_jobs=-1)
    return scores
```

```python
# load the dataset
X, y = load_dataset(full_path)
# summarize the loaded dataset
print(X.shape, y.shape, Counter(y))
# define the reference models
steps = [
    ("t1", StandardScaler()),
    ("t2", PowerTransformer()),
    (
        "m",
        LogisticRegression(solver="lbfgs", penalty=None, class_weight="balanced"),
    ),
]
models = [DummyClassifier(strategy="constant", constant=1), Pipeline(steps=steps)]
names = ["Dummy", "LR"]
# evaluate the models
baseline_performance = []
for model, name in zip(models, names):
    scores = evaluate_model(X, y, model)
    baseline_performance.append(mean(scores))
    # summarize performance
    print("Mean F1 (%s): %.3f (%.3f)" % (name, mean(scores), std(scores)))
```

### Try out some models

Utility function.

```python
# load the dataset
def load_dataset(full_path):
    # load the dataset as a numpy array
    data = read_csv(full_path, header=None)
    # retrieve numpy array
    data = data.values
    # split into input and output elements
    X, y = data[:, :-1], data[:, -1]
    # label encode the target variable to have the classes -1 (outlier) and
    # +1 (inlier)
    y = LabelEncoder().fit_transform(y)
    y = -2 * y + 1
    return X, y


# evaluate a model
def evaluate_model(X, y, model):
    # define evaluation procedure
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    metric = make_scorer(f1_score, pos_label=-1)
    # evaluate model
    scores = cross_val_score(model, X, y, scoring=metric, cv=cv, n_jobs=-1)
    return scores


# define models to test
def get_models():
    models, names = list(), list()
    # One-class SVM
    steps = [("t", StandardScaler()), ("m", SGDOneClassSVM(random_state=1))]
    models.append(Pipeline(steps=steps))
    names.append("SVM")
    # One-class SVM + Power
    steps = [
        ("t1", StandardScaler()),
        ("t2", PowerTransformer()),
        ("m", SGDOneClassSVM(random_state=1)),
    ]
    models.append(Pipeline(steps=steps))
    names.append("SVM-Power")
    # Isolation forest
    models.append(IsolationForest(n_jobs=-1, random_state=1))
    names.append("IF")
    # Isolation forest + Power
    steps = [
        ("t", PowerTransformer()),
        ("m", IsolationForest(n_jobs=-1, random_state=1)),
    ]
    models.append(Pipeline(steps=steps))
    names.append("IF-Power")
    # Minimum covariance determinant
    models.append(EllipticEnvelope(support_fraction=0.8, random_state=1))
    names.append("MCD")
    # Minimum covariance determinant + Power
    steps = [
        ("t", PowerTransformer()),
        ("m", EllipticEnvelope(support_fraction=0.8, random_state=1)),
    ]
    models.append(Pipeline(steps=steps))
    names.append("MCD-Power")
    # Local outlier factor
    steps = [
        ("t", StandardScaler()),
        ("m", LocalOutlierFactor(n_jobs=-1, novelty=True)),
    ]
    models.append(Pipeline(steps=steps))
    names.append("LOF")
    # Local outlier factor + Power
    steps = [
        ("t1", StandardScaler()),
        ("t2", PowerTransformer()),
        ("m", LocalOutlierFactor(n_jobs=-1, novelty=True)),
    ]
    models.append(Pipeline(steps=steps))
    names.append("LOF-Power")
    return models, names
```

Note that the models do not require tuning, so the cross-validation error
is a honest estimate of the generalization error of these models
(in this case measured in terms of F1 score).

```python
# load the dataset
X, y = load_dataset(full_path)
# define models
models, names = get_models()
results = list()
# evaluate each model
for i in range(len(models)):
    # evaluate the model and store results
    scores = evaluate_model(X, y, models[i])
    results.append(scores)
    # summarize and store
    print(">%s %.3f (%.3f)" % (names[i], mean(scores), std(scores)))
# plot the results
pyplot.boxplot(results, labels=names, showmeans=True)
pyplot.axhline(baseline_performance[0])
pyplot.axhline(baseline_performance[1])
pyplot.show()
```
