### Create a baseline model

Some utility functions.

Note that we use stratified cross-validation to ensure that the distribution
of class labels in each fold is the same as the distribution of the class
labels in the full dataset.

```python
# load the dataset
def load_dataset(full_path):
    # load the dataset as a numpy array
    data = read_csv(full_path, header=None)
    # retrieve numpy array
    data = data.values
    # split into input and output elements
    X, y = data[:, :-1], data[:, -1]
    # label encode the target variable to have the classes 0 and 1
    y = LabelEncoder().fit_transform(y)
    return X, y


# evaluate a model
def evaluate_model(X, y, model):
    # define evaluation procedure
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    # evaluate model
    scores = cross_val_score(model, X, y, scoring="roc_auc", cv=cv, n_jobs=-1)
    return scores
```

```python
# load the dataset
X, y = load_dataset(full_path)
# summarize the loaded dataset
print(X.shape, y.shape, Counter(y))
# define the reference model
model = DummyClassifier(strategy="stratified", random_state=1)
# evaluate the model
scores = evaluate_model(X, y, model)
baseline_performance = mean(scores)
# summarize performance
print("Mean ROC AUC: %.3f (%.3f)" % (mean(scores), std(scores)))
```

Note that we are using a `DummyClassifier` that predicts a random class label
in proportion to the observed frequency of each label.

This is a reasonable baseline because it gives us AUC = 0.50, which is the
no-skill classifier in this case.

### Try out some models

Utility function.

```python
# load the dataset
def load_dataset(full_path):
    # load the dataset as a numpy array
    data = read_csv(full_path, header=None)
    # retrieve numpy array
    data = data.values
    # split into input and output elements
    X, y = data[:, :-1], data[:, -1]
    # label encode the target variable to have the classes 0 and 1
    y = LabelEncoder().fit_transform(y)
    return X, y


# evaluate a model
def evaluate_model(X, y, model):
    # define evaluation procedure
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    # evaluate model
    scores = cross_val_score(model, X, y, scoring="roc_auc", cv=cv, n_jobs=-1)
    return scores


# define models to test
def get_models():
    models, names = list(), list()
    # LR
    models.append(LogisticRegression(solver="lbfgs", penalty=None))
    names.append("LR")
    # SVM
    steps = [("t", StandardScaler()), ("m", SVC(gamma="scale"))]
    models.append(Pipeline(steps=steps))
    names.append("SVM")
    # Bagging
    models.append(BaggingClassifier(n_estimators=100, n_jobs=-1))
    names.append("BAG")
    # RF
    models.append(RandomForestClassifier(n_estimators=100, n_jobs=-1))
    names.append("RF")
    # GBM
    models.append(GradientBoostingClassifier(n_estimators=100))
    names.append("GBM")
    return models, names
```

Note that the models do not require tuning, so the cross-validation error
is a honest estimate of the generalization error of these models
(in this case measured in terms of the ROC AUC).

```python
# load the dataset
X, y = load_dataset(full_path)
# define models
models, names = get_models()
results = list()
# evaluate each model
for i in range(len(models)):
    # evaluate the model and store results
    scores = evaluate_model(X, y, models[i])
    results.append(scores)
    # summarize and store
    print(">%s %.3f (%.3f)" % (names[i], mean(scores), std(scores)))
# plot the results
pyplot.boxplot(results, labels=names, showmeans=True)
pyplot.axhline(baseline_performance)
pyplot.show()
```

All of these models are skillful.

### Try including training weights

Let's see if we can improve this further by adding weights.

```python
# define models to test
def get_models():
    models, names = list(), list()
    # LR
    models.append(LogisticRegression(solver="lbfgs", penalty=None, class_weight="balanced"))
    names.append("LR")
    # SVM
    steps = [
        ("t", StandardScaler()),
        ("m", SVC(gamma="scale", class_weight="balanced", random_state=1)),
    ]
    models.append(Pipeline(steps=steps))
    names.append("SVM")
    # Bagging
    models.append(
        BaggingClassifier(
            estimator=DecisionTreeClassifier(class_weight="balanced"),
            n_estimators=100,
            n_jobs=-1,
            random_state=1,
        )
    )
    names.append("BAG")
    # RF
    models.append(
        RandomForestClassifier(
            class_weight="balanced", n_estimators=100, n_jobs=-1, random_state=1
        )
    )
    names.append("RF")
    return models, names
```

```python
# define models
models, names = get_models()
results = list()
# evaluate each model
for i in range(len(models)):
    scores = evaluate_model(X, y, models[i])
    results.append(scores)
    # summarize and store
    print(">%s %.3f (%.3f)" % (names[i], mean(scores), std(scores)))
# plot the results
pyplot.boxplot(results, labels=names, showmeans=True)
pyplot.axhline(baseline_performance)
pyplot.show()
```

It appears that introducing training weights helped for
support vector machine classifier in this case. The bagging classifier and the
random forest do not appear to have benefitted from the use of training weights.
The performance of logistic regression is unaffected.
