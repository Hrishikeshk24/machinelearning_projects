First, import the dataset and create a **80-20** train-test split. Use **886** as the random seed throughout your code. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.tree import export_graphviz
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from graphviz import Source
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0,1],[0,1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
```

```python
pd.options.display.max_columns = None
```

#### Part (a) [15pts]

Construct a linear regression model on the training set. You can use all the variables in he dataset (that is, do not worry about issues of multi-collinearity or variable selection). What is the in-sample and out-of-sample R-squared for this model?


```python
df = pd.read_csv("ames.csv")
df.head()
```

```python
#df.dropna(axis = 0, how = 'any', inplace = True)
```

```python
#df_oneHot = pd.get_dummies(df, drop_first=True)

Y = df["SalePrice"]
X = df.drop("SalePrice", axis=1)

X_encoded = pd.get_dummies(X, drop_first=True)
X_encoded
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y, test_size=0.2, random_state=886)
```

```python
lin_reg = LinearRegression().fit(X_train, Y_train)
inSampler2 = lin_reg.predict(X_train)
r2_score(Y_train,inSampler2)
```

```python
lin_reg.intercept_
```

In-sample R^2 for this model is approximately 0.90 or 90%. 

```python
test_lin_reg = lin_reg.predict(X_test)
outSampler2 = r2_score(Y_test, test_lin_reg)
outSampler2
```

Out-sample r^2 for the linear regression model is approximately 0.864 or 86.4%. 

#### Part (b) [15pts]

Train a CART model with max_depth set to 3 and without changing any other default parameters. Use a random_state value equal to 886.

```python
treemod = DecisionTreeRegressor(max_depth=3, random_state=886)
treemod.fit(X_train, Y_train)

export_graphviz(
        treemod,
        out_file="ames.pdf",
        feature_names=X_train.columns,
        impurity=False,
        rounded=True,
        filled=True
    )
```

```python
Source.from_file("ames.pdf")
```

```python
inSampler2_tree = treemod.score(X_train, Y_train)
inSampler2_tree
```

```python
test_tree = treemod.predict(X_test)
outSampler2_tree = r2_score(Y_test, test_tree)
outSampler2_tree
```

#### Part (d) [25pts]

Use **10-fold** cross-validation to select a good max_depth value for your tree and then fit your final model. Consider max_depth values from 2 to 10. What is the in-sample and out-of-sample R-squared for the selected model (fitted on the entire train set)? Use "r2" as the scoring method.

```python
treemod.get_params()
```

```python
parameter_grid = [
  {'max_depth': [2, 3, 4, 5, 6, 7, 8, 9, 10]} 
]
```

```python
cv_grid = GridSearchCV(treemod, parameter_grid, scoring="r2", cv = 10)
```

```python
cv_grid.fit(X_train, Y_train)
```

```python
pd.DataFrame(cv_grid.cv_results_)
```

According to my cross validation model above, the best max depth to use for our CART model is a max depth of 5. 

```python
treemod_cv = DecisionTreeRegressor(max_depth=6, random_state=886)
treemod_cv.fit(X_train, Y_train)

export_graphviz(
        treemod_cv,
        out_file="ames_cv.pdf",
        feature_names=X_train.columns,
        impurity=False,
        rounded=True,
        filled=True
    )
```

```python
Source.from_file("ames_cv.pdf")
```

```python
inSampler2_tree_cv = treemod_cv.predict(X_train)
r2_score(Y_train, inSampler2_tree_cv)
```

```python
test_tree = treemod_cv.predict(X_test)
outSampler2_tree_cv = r2_score(Y_test, test_tree)
outSampler2_tree_cv
```

#### Part (e) [25pts]

Construct a random forest model with **1000** trees. Leave other hyperparameters at the default values. What is the in-sample and out-of-sample R-squared for the selected model? Again use a random_state equal to 886.

```python
rf_mod = RandomForestRegressor(n_estimators = 1000, random_state=886) # Letting max_feature parameter to be the default value for this model. 
rf_mod.fit(X_train, Y_train)
```

```python
print("Random Forest Train R2: ", rf_mod.score(X_train, Y_train))
print("Random Forest Test R2: ", rf_mod.score(X_test, Y_test))
```
