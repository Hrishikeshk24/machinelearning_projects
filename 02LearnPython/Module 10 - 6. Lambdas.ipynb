```python
add = lambda x, y: x + y
```

```python
sub = lambda x, y, z: x - y - z
```

```python
mul = lambda x, y: x * y
```

```python
div = lambda x, y: x / y
```

```python
def calculate(x, y, operation="add"):
    
    if operation == "add":
        return add(x, y)
    if operation == "sub":
        return sub(x, y)
    if operation == "mul":
        return mul(x, y)
    if operation == "div":
        return div(x, y)
```

```python
calculate(100, 4, operation="sub")
```

```python
calculate(100, 4)
```

```python
calculate(100, 4, operation="mul")
```

```python
def calculate(*args, operation="add"):
    
    if operation == "add":
        return add(*args)
    if operation == "sub":
        return sub(*args)
    if operation == "mul":
        return mul(*args)
    if operation == "div":
        return div(*args)
```

```python
calculate(100, 4, 5, operation="sub")
```

```python
calculate(100, 3, operation="mul")
```

```python
calculate(100, 3)
```
