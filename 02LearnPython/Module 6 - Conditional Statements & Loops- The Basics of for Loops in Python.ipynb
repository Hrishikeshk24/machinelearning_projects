# Iterating over elements in a tuple and a dictionary

```python
for country in 'India', 'US', 'Switzerland':
    print(country)
```

```python
## when we have values separated by commas, even if they are not placed within square brackets, python asusmes that this
# is to be treated as a sequence where elements are separated by commas.
```

```python
dogs=('Pug', 'Doberman', 'Golden Retriever')
## tuple
```

```python
for dog in dogs:
    print('it\'s a', dog )
```

```python
dog_weight =(('Pug',20),
            ('Doberman', 80),
            ('Golden Retriever', 55))

for i, (dog, weight) in enumerate (dog_weight):
    print ('The dog at index %s is %s and it weighs %s' %(i, dog, weight))
```

```python
## an element at each iteration will be an individual inner tuple.
## Note that we are not iterating directly over dog_weight instead use enumerate function. What is this?
## We will see that at each iteration it will return not only the individual element, which is a tuple but also an index for that
## element within the object which we are iterating over.
```

```python
friends = ['John', 'Sam', 'Mike']

for friend in friends:
    print('Hapy New year,', friend)
```

# The else block of a FOR LOOP

```python
numbers = [4,6,9,10]

for num in numbers:
    quotient = num//2
    print(quotient, 'is the quotient of',num,'/2')
```

```python
mixed_list = [145, 10.5, 1+3j, True, "Python", (0,1), [2,-5], {"class": "V", "Section":"A"}, {1,2,3}]
```

```python
for item in mixed_list:
    print('Type of ', item, 'is', type(item))
```

```python
num_list = [2,4,6,8]

for val in num_list:
    square = val ** 2
    print('Square of', val, 'is', square)
```

```python
string = input('enter a string: ')

count = 0

for i in string:
    count +=  1
print(count)
```

```python
## Using else block in a for loop - It means that the execution of code will enter the else block after 
## all of the iterations of the FOR loop have been completed successfully. So the else block will be executed everytime 
##with some exceptions.
```

```python
num =[2,3,5,7]

for i in num:
    print(i)
else:
    print('No items left in the list')
```

```python
## So when will the else block be not executed with FOR Loop?
## We could have an error during the processing of each element within the FOR loop, or we could also exit prematurely from
## the FOR loop before it gets to complete all iteration.
```

```python
char = ['a', 'b', 'c', 'd']

for i in char:
    print(i)
else:
    print('No more char')
```

```python
char = {'a', 'b', 'c', 'd'}
## set - order of the elements will not be maintained as it is the set.

for i in char:
    print(i)
else:
    print('No more char')
```
