# Performing shallow copies of lists

```python
# As far as indexing and slicing operations go, lists and strings are similar. 
# As far as copying function, these are very different as strings are immutable and lists are not.
```

```python
name = ['ram', 'pintu','bheem']
```

```python
name
```

```python
new_names= name
print('copy:',new_names)
print('Original:',name)
```

```python
name[1]
```

```python
new_names[1]='appa'
print('copy:',new_names)
print('Original:',name)
```

```python
# Observe that both name and new_names lists have been updated.
# How dd this happen? Because when we used an assignment operator, we created a shallow copy of 'name' list. Both of the vars
# name and new_names refer to the same list under the hood in the memory. This is the same list.
# Notice that this was not the case with STRINGS.
```

```python
import copy
```

```python
companies = ['fb','amazon','tsla']
```

```python
companies
```

```python
new_comp = copy.copy(companies)
print('companies:',companies)
print('new_comp:',new_comp)
```

```python
new_comp[1]
```

```python
new_comp[1]='msft'
```

```python
print('companies:',companies)
print('new_comp:',new_comp)
```

```python
# this means, under the hood, companies and new_comp are two separate lists.
# copy.copy(list) creates a deep copy of a list. Deep copy means an entrely new list will be created in the memory 
#and changes to the copy will not affect the original list.
```

```python
companies[1]='aapl'
```

```python
print('companies:',companies)
print('new_comp:',new_comp)
```

```python
## So assignment operator on the list created a Shallow copy.
## Copy module created a Deep copy.
```

```python
old_list=[[1,2,3],[4,5,6],[7,8,9]]
old_list
```

```python
new_list = copy.copy(old_list)
print('old_list:',old_list)
print('new_list:',new_list)
```

```python
old_list[1][1]
```

```python
old_list[1][1]='five'
print('old_list:',old_list)
print('new_list:',new_list)
```

```python
## IMPORTANT
##Copy function we passed created a deep copy of the outer list and a shallow copy of the inner list.
```

```python
new_list[2].append('Ten')
print('old_list:',old_list)
print('new_list:',new_list)
```

```python
## the copy function only created a shallow copy of all the inner list, which means they refer the same portion in Python memory.
```

```python
old_list.append([10,11,12])
print('old_list:',old_list)
print('new_list:',new_list)
```

```python
## Note that we are appending the outer list and we see that the updates have been applied only to the old_list.
## Copy func creates deep copy of the outer list , so changes you make to the outer list will not be reflected in the copy.
## But any change made to the inner nested list will be reflected both in the copy and the original list.
```
