```python
class Student:
    
    def __init__(self,firstname,lastname):
        self.fname = firstname
        self.lname = lastname
        self.email = firstname + "."+ lastname + "@xyz.com"
        
    def uppercase(self):
        self.fname = self.fname.upper()
        self.lname = self.lname.upper()
        
    def fullname(self):
        return '{} {}'.format(self.fname, self.lname)
        
```

## CREATE CLASSES TO REPRESENT A STUDENT

```python
class Student:
    
    def __init__(self, name, gpa):
        self.__name = name
        self.__gpa = gpa
        self.__clubs =[]
        self.__active = True
        
    def get_name(self):
        return self.__name
    
    def get_gpa(self):
        return self.__gpa
    
    def change_name(self, name):
        self.__name = name
        
    def change_gpa(self, gpa):
        self.__gpa = gpa
        
    def change_name_gpa(self, name, gpa):
        self.change_name(name)
        self.change_gpa(gpa)
        
    def add_clubs(self,club):
        self.__clubs.append(club)
        
    def update_status(self, status):
        self.__active = status
        
    def print_details(self):
        print('My name is %s and my gpa is %s and I belong to the clubs %s and my active status is %s' %(self.__name, self.__gpa, self.__clubs, self.__active)) 
               
              
    
```

```python
s1 = Student('Hrishi', 3.0)
```

```python
s1
```

```python
s1.print_details()
```

```python
s1.get_name()
```

```python
s1.get_gpa()
```

```python
s1.change_name_gpa('Shru', 4.0)

s1.print_details()
```

```python
s1.add_clubs('Liverpool')
s1.print_details()
```

```python
s1.add_clubs('Manchester United')
s1.print_details()
```

```python
s1.update_status(False)

s1.print_details()
```

```python
s2 = Student('Hrishi', 3.5)

s2.print_details()
```

```python
s1.get_name()
```

```python
s2.get_name()
```

```python
import math
class Circle:
    
    def __init__(self, radius):
        self.__rad = radius
        
    def calc_area_per(self):
        self.__area = math.pi * self.__rad * self.__rad
        self.__per = 2 * math.pi * self.__rad
        
    def print_details(self):
        print('The area of the circle is %s and the perimeter of the circle is %s'  %(self.__area, self.__per))
    
    
```

```python
c1=Circle(12)
```

```python
c1.calc_area_per()
c1.print_details()
```
