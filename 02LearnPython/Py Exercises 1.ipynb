```python
## Write a Python program to compute the greatest common divisor (GCD) of two positive integers.
## REVISIT THIS
def gcd(x, y):
   gcd = 1   
   if x % y == 0:
       return y   
   for k in range(int(y / 2), 0, -1):
       if x % k == 0 and y % k == 0:
           gcd = k
           break 
   return gcd
print("GCD of 12 & 17 =",gcd(17, 17))
print("GCD of 4 & 6 =",gcd(4, 6))
print("GCD of 336 & 360 =",gcd(336, 360))
```

```python
## Write a Python program to get height and width of the console window.

def terminal_size():
    import fcntl, termios, struct
    th, tw, hp, wp = struct.unpack('HHHH',
        fcntl.ioctl(0, termios.TIOCGWINSZ,
        struct.pack('HHHH', 0, 0, 0, 0)))
    return tw, th

print('Number of columns and Rows: ',terminal_size())
```

```python
## Write a Python program to get execution time for a Python method.

import time
def sum_of_n_numbers(n):
    start_time = time.time()
    print(start_time)
    s = 0
    for i in range(1,n+1):
        s = s + i
    end_time = time.time()
    print(end_time)
    return s,end_time-start_time

n = 5
print("\nTime to sum of 1 to ",n," and required time to calculate is :",sum_of_n_numbers(n))
```

```python
## Write a Python function to find the maximum and minimum numbers from a sequence of numbers

def minimum(lst):
    mini = lst[0]
           
    for i in lst:
        if mini > i:
            mini = i
        else:
            continue
    return mini           
           
```

```python
minimum([3,2,4,5,6,7,87,84,3,435,435656,-2, 456,435,3534,23,41,11,1,0, -1])
```

```python
def maximum(lst):
    maxi = lst[0]
           
    for i in lst:
        if maxi < i:
            maxi = i
        else:
            continue
    return maxi 
```

```python
maximum([3,2,4,5,6,7,87,84,3,435,435656,-2, 456,435,3534,23,41,11,1,0, -1, 1000000])
```

```python
def max_min(data):
  l = data[0]
  s = data[0]
  for num in data:
    if num> l:
      l = num
    elif num< s:
        s = num
  return l, s

print(max_min([0, 10, 15, 40, -5, 42, 17, 28, 75]))
```

```python
## Write a Python function that takes a positive integer and returns the sum of the cube of 
## all the positive integers smaller than the specified number.

def sum_cube(x):
    n = 1
    sum_cube = 0
    while n < x:
        sum_cube = sum_cube + (n**3)
        n+=1
    return sum_cube
```

```python
## Write a Python function that takes a sequence of numbers and determines 
## whether all the numbers are different from each other

def test_distinct(data):
    
    if len(data) == len(set(data)):
        return True
    else:
        return False
```

```python
print(test_distinct([1,5,7,9]))
print(test_distinct([2,4,5,5,7,9]))
```
