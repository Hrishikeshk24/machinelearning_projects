# The Continue statement - Part 1

*With the break statement, we can check for a specific condition during an iteraion and if the condition is met, we can break out of the FOR loop.
Until that condition is met though, we will continue performing certain steps within the body of the FOR loops.
The break stmnt though, represents a somewhat extreme course of action.
Consider the case where we are iterating over a sequence. And if we happen to find a value, we don't want to break out of the FOR loop,
but we simply wish to skip that particular iteration and then move on to the next one without executing any of the actions
defined within the for loop body. This is where continue stmnt come into picture.*


```python
for letter in 'string':
    if letter == 'i':
        continue
    print(letter)
print('The end')
```

```python
for letter in 'string':
    if letter == 'i':
        break
    print(letter)
print('The end')
```

```python
my_string = 'Students'

for letter in my_string:
    
    if (letter == 'e' or letter =='s'):
        continue
    else:
        print(letter)
```

```python
for i in range(1,10):
    
    if i%2==0:
        print('Is %s an even number?' %i)
    else:
        continue
    print('Yes')
```

```python
for i in range(1,10):
    
    if i%2==0:
        print('Is %s an even number?' %i)
        print('Yes')
    else:
        continue
    
```

```python
for num in range(1,10):
    if num % 2 == 0:
        print('Found an even number:', num)
        
        continue
    print(num)
```

```python
for num in range(1,10):
    if num % 2 == 0:
        print('Found an even number:', num)
        
        continue
    else:
        print(num)
```

```python
num_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31 ]

months = ['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

new_dict= dict() # initializing a dictionary.


# to load values into the dictionary, use the zip function.This func will combine two lists in order to form a list of tuples.
# By zipping the lists, months, and num_days (all same length), we will be creaeting a list of tuples where within each tuple, 
# we will have the name of the month and number of days in that month.
# We set up a for loop in order to iterate over each of the tuples in that list of tuples and the individual elements within the tuple
# will be assigned to the variables i and j for each iteration.
# Now the first thing we do within each iteartion of the FOR loop is to add the values i and j into a dictionary.
# where i becomes the key and j, the value. So this is how we load our dictionary.
# Make note of the use of update function for the dictionary here
# where if the key value pair happens to be new , then it will be added to the dictionary and if the key already exists , its coresponding
# value will be updated.



for i, j in zip(months, num_days):
    new_dict.update({i:j})
    
    if new_dict[i] < 29:
        continue
    
    else:
        print(i, 'month has', new_dict[i], 'days')
```

```python
for letter in "string":
    
    if letter =='r':
        pass

    print(letter)
```
