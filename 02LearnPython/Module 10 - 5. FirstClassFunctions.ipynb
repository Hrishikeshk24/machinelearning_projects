## Functions as Arguments - Variable Arguments

```python
def calculate(*args, fn):
    return fn(*args)
```

```python
calculate(10, fn=diameter_circle_fn)
```

```python
calculate(10, fn=perimeter_circle_fn)
```

```python
calculate(10, fn=area_circle_fn)
```

```python
def area_rectangle_fn(length, breadth):
    
    return length * breadth
```

```python
def perimeter_rectangle_fn(length, breadth):
    
    return 2 * (length + breadth)
```

```python
calculate(20, 40, fn=area_rectangle_fn)
```

```python
calculate(20, 40, fn=perimeter_rectangle_fn)
```

```python
calculate(10, fn=area_rectangle_fn)
```

## Functions as Return Values

```python
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    return a / b
```

```python
def get_function(operator='+'):
    
    if operator == "+":
        return add ## these return values are functions

    if operator == "-":
        return sub
    
    if operator == "*":
        return mul 
    
    if operator == "/":
        return div 
```

```python
func = get_function()

func
```

```python
func(3, 4)
```

```python
func = get_function('*')

func
```

```python
func(3, 4)
```

```python
calc_dictionary = {
    '+': add,  ## these values are functions
    '-': sub,
    '*': mul,
    '/': div
}
```

```python
func = calc_dictionary['/']

func
```

```python
func(12, 4)
```

```python
func = calc_dictionary['*']

func
```

```python
func(10, 20)
```

```python
calc_dictionary['*'](100, 3)
```

```python
calc_dictionary['+'](100, 3)
```
