# Argument passing by Value

```python
fruits_list = ['Apple', 'Banana', 'Orange']
```

```python
def change_list(fruits_list):
    
    fruits_list = ['Kiwi', 'Strawberry']
    print('Inside the function:', fruits_list)
    
change_list(fruits_list)
print()
print('Outside the function:', fruits_list)
```

```python
fruits_list = ['Mango', 'Grapes']
```

```python
change_list(fruits_list)
```

```python
fruits_list
```

```python
##Passingby value means when you re assign a list within a function , that reassignment or creation of the new list does not 
## affect the list outside the function. This is true for tuples and dicts as well.
```

```python
def change_list(fruits_list):
    
    fruits_list = "This is not even a list"  # changed to a string. Same for Tuples and dicts
    print('Inside the function:', fruits_list)
    
change_list(fruits_list)
print()
print('Outside the function:', fruits_list)
```

```python
## you can see that the fruits_list has been passed in by value.The value of the original list remains unchanged even though 
## you have changed its value within the function. This is waht passing by value means.
```

```python
car_speeds = {
    "camry":120,
    "Accord": 130,
    "Cayenne": 250,
    "Mini":100
}
```

```python
def change_dict(car_speeds):
    
    car_speeds = {
        "Mustang" : 200
    }
    
    print('Inside the function:', car_speeds)
```

```python
change_dict(car_speeds)
```

```python
car_speeds
```

```python
## This has not affected the dict defined outside the function
```

```python
## For both primitve and complex data types, when you re-assign the values of these data types inside the function, you'll
## find the vars outside the function remain unaffected. This is what passing by value means.
```

# Argument passing by reference

```python
fruits_list = ['Apple', 'Grapes', 'Mango', 'Banana']
```

```python
def change_list(fruits_list):
    
    fruits_list[0] = 'Kiwi'
    print('Inside the function:', fruits_list)
    
change_list(fruits_list)
print()
print('Outside the function:', fruits_list)
```

```python
## Observe that the list has been updated both inside and outside the function. Global var has also been updated.
## why did the fruits list which was declared outside the func also changed?
## Because we accept it as an input into the change_list function and modified the actual contents of the list.
## We just didn't re-assign the list, we changed the content of the list.
```

```python
## If we had t like fruits_list = ['Kiwi']  - new list with just Kiwi in it. This is reassigning the 
## fruits_list inside the function. In this case, the original list remains unchnaged and its located somewhere in memory.
## But in the above example, we are modifying the list itself (contents).
```

```python
fruits_list = ['Apple', 'Grapes']
```

```python
def change_list(fruits_list):
    
    fruits_list = ['Kiwi']
    print('Inside the function:', fruits_list)
    
change_list(fruits_list)
print()
print('Outside the function:', fruits_list)
```

```python
fruits_list
```

```python
def change_list(fruits_list):
    
    fruits_list[0] = 'Kiwi'
    print('Inside the function:', fruits_list)
    
change_list(fruits_list)
print()
print('Outside the function:', fruits_list)
```

```python
fruits_list
```

```python
fruits_list = ['Apple', 'Grapes', 'Mango', 'Banana']

def change_list(fruits_list):
    
    fruits_list.append('Kiwi')
    print('Inside the function:', fruits_list)
    
change_list(fruits_list)
print()
print('Outside the function:', fruits_list)
```

```python
## We modified the list and hence there is change in the global var also.
```

```python
fruits_list = ['Apple', 'Grapes', 'Mango', 'Banana']

def change_list(fruits_list):
    
    fruits_list.remove('Apple')
    print('Inside the function:', fruits_list)
    
change_list(fruits_list)
print()
print('Outside the function:', fruits_list)
```

```python
## Both of the local and global vars reference the same list in memory, the change is reflected inside and outside the func.
```

```python
## This demonstrates that when we work with complex data types, the argument passing is by REFERENCE.
```

```python
fruits_list = ['Apple', 'Grapes', 'Mango', 'Banana']

def sort_items_remove_last(my_list):
    
    my_list.sort()
    my_list.pop()
    
    print('Inside the function:', my_list)
```

```python
sort_items_remove_last(fruits_list)
```

```python
fruits_list
```

```python
## Observe that the varibale names are different. We are using my_list in the function definition. but outside the func, 
## we have fruits_list. But still both local and global vars are affected.
## It doesn't matter what the name of the variable is. Both reference the same list in the memory.

## This proves to us that the contents of the complex data type, such as list can be changed from within the function, and the
## same is true for dicts as well. Tuples are not mutable, so we will ignore tuples for now.
```
