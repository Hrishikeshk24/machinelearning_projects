```python
def find_first_cap_letter(some_string):
    
    cap_letter = None
    
    for ch in some_string:
        if ch.upper() == ch and ch != ' ':
            cap_letter = ch
            break
            
    if cap_letter is None:
        return 'No Cap letters found'
    else:
        return 'First Capital letter is '+  cap_letter
```

```python
## So how deo we get this func to accept any no. of args like the built-in print function does?

## There is a special concept for this.
```

```python
def print_fn(*args): # this is the syntax Python uses to indicate that you have a function that can accept variable number of args.
                     # an aseterisk and name of the variable     
    
    args_type= type(args)
    print(args_type)
    
    print(args)
```

```python
print_fn()
```

```python
## It's an empty tuple because we invoked the func without any input args.

##The fact that all of the inputs are recived in the form of a tuple is what gives this function the flexibility to accept 
##variable number of args.
```

```python
print_fn('bob')
```

```python
print_fn('bob', 'Peter')
```

```python
print_fn('bob', 'Peter', 'Mark')
```

```python
## Invoke as many args as possible and use this custome function to print just like the built-in function.

## The presence of an astrisk indicates to Python that there can be any num of args. Just pack them all into the tuple and
## pass a single tuple into this python func.
```

```python
## What if all the elements that you want to pass is already in the list or tuple format?
```

```python
name_list = ['mike', 'george', 'john',' mark']

print_fn(name_list)
```

```python
## Here the list is considered a single entity whihc is then packed into a tuple. 
## You see that the list is the first field in the tuple.
## Python has considered the entire list as the single element, and its the first field in the tuple that is passed in.


## What if we want the individual elements that make up the name_list to be separate input args to your print fun?
## You will ask Python to unpack the name_list and then pass this unpacked list to the print function.
## Once inside th print func, python will repack the individual elements of the list as individual elements of the tuple 
## that is passed into the print_fn function, the args tuple.
```

```python
print_fn(*name_list)
```

```python
## Here you see all the names are diff fields in the tuple.
```

# Variable length args - **kwargs keyword

```python
## Another way to specify var length args as an input to a function. If you want your var len arg to be passed in in the form of
## keywords, yu will specify the double asterisk in the front of your input arg. A common var is kwargs.
```

```python
def student_details(**kwargs):
    print(type(kwargs))
    print(kwargs)
```

```python
def student_details1(*args):
    print(type(args))
    print(args)
```

```python
## when we use double asterisk, python packs var len arg into a dict not a tuple.
```

```python
student_details() ## since it is a var len arg, no need to specify any values and it works fine.
```

```python
student_details1()
```

```python
student_details(name="john") ## Python will pack these args into a dict. Here kwargs is a dic with 1 key value pair. So we 
                               ## should always specify this in the key value pair format.
```

```python
student_details(name = "john", age =18)
```

```python
def student_details(**details): ## Here our var len arg is called details
    
    for key, value in details.items():
        print(key, value)
```

```python
student_details(name = "john", age =18)
```

```python
## What if the values you want to use to invoke the func is already in the dict format?
```

```python
details_dict = {'name': 'John', 'age':18, 'college': 'Harvard'}
```

```python
student_details(details_dict)
```

```python
## This will not work as it is expecting key word args and details dict is passed in as a pos arg. So it should be like this.

student_details(**details_dict)
```

```python
def student_details(**details):
    
    if 'name' in details:
        print('Name:', details['name'])
        
    if 'age' in details:
        print('Age:', details['age'])
        
    if 'college' in details:
        print('College:', details['college'])
```

```python
student_details(name='babe')
```

```python
student_details(name='babe', college = 'Boston')
```

```python
student_details(name='babe', college = 'Boston', age =18)
```

```python
## You can have the same func excpet var len args as positional args as well as keyword args.
```

```python
def students_in_college(*students, **college_details):  ##students is a positional var len arg
                                                        ## college_details is a kw var len arg   
    print("Students____")
    for s in students:
        print (s)
        
    print()
    
    print('College_details____')
    for key, value in college_details.items():
        print(key, value)
```
